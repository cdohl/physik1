## 4. Dynamik der Punktmasse
### 4.1 Kraftbegriff und Inertialsystem

#### 1. Newtonsche Gesetz

<b>In einem Intertialsystem bewegt sich ein Körper geradlinig mit konstanter Geschwindigkeit, wenn keine Kraft auf den Körper wirkt.</b>


Ein Inertialsystem ist ein nicht beschleunigtes Koordinatensystem. *Ist der Hörsaal ein Inertialsystem?*

Betrachten wir die Radialbeschleunigung auf der Erde:

<img src="figures/4.1_1.png" width=250px>

Setzen wir den Erdradius von ca. $R_e\approx 6400\,$km und die Winkelgeschwindigkeit von $\omega=2\,\pi/(24*3600)\,$s$^{-1}$ ein ist die Radialbeschleunigung am Äquator maximal  $a_r=0.034\,$m/s$^2$.
Das ist zwar deutlich geringer als die Gravitationsbeschleunigung, dennoch ein beschleunigtes Koordinatensystem. Obwohl der Hörsaal damit kein exaktes Inertialsystem ist, ist er doch eine gute Näherung dafür.


#### 2. Newtonsche Gesetz

Nehmen Sie eine Feder und ziehen die Feder um eine feste Strecke aus. An der Feder hängt eine Masse $m_1$. Die Feder zieht an der Masse $m_1$ und beschleunigt diese auf $a_1$. Wenn Sie jetzt bei gleicher Auslenkung eine andere Masse $m_2$ befestigen, zieht die Feder an Masse $m_2$ mit der Beschleunigung $a_2$. 

Experimentell ist nachgewiesen worden, dass

$$m_1\,a_1=m_2\,a_2$$

<img src="figures/4.1_2.png" width=350px>

Das Produkt aus $m$ und $a$ ist die Definition für Kraft. Das 2. Newtonsche Gesetz lautet:

<b> Eine Kraft auf einen Körper beschleunigt diesen in Richtung der Kraft und hat den Betrag $m\,a$.</b>

$$\vec{F}=m\,\vec{a}$$

Achtung, das gilt nur für Körper mit konstanter Masse. Die Einheit von $F$

$$\left[F\right]=\mathrm{kg}\,\frac{\mathrm{m}}{\mathrm{s}^2}=\mathrm{N}$$ 

N steht für Newton.

Die Gravitationskraft ist demnach

$$\vec{F}_g=m\,\vec{g}$$

wobei $\vec{g}$ die Erdbeschleunigung ist. Wenn Sie einen blauen Ball in der Hand halten wie im Bild gezeigt, bewegt sich der Ball nicht. Deshalb muss die Kraft auf den Ball 0 sein. Neben der Gravitationskraft $\vec{F}_g$ wirkt noch die Kraft der Hand $\vec{F}_{\mathrm{Hand}}$.

<img src="figures/4.1_3.png" width=300px>

Die Gesamtkraft ist die Summe der Kräfte und damit 0

$$\vec{F}_g+\vec{F}_{\mathrm{Hand}}=0\quad $$

und damit ist 
$$\vec{F}_{\mathrm{Hand}}=-m\vec{g}$$.

#### Beispiel Kräftegleichgewicht

Die Mase $m$ hängt in Ruhe. 

<img src="figures/4.1_4.png" width=400px>

Daher ist die resultierende Beschleunigung 0:

$$\sum_i\vec{F}_i=m\vec{a}=0$$

also
$$\vec{F}_1+\vec{T}_2+\vec{T}_3 =0$$

Wir betrachten die $x$- und $y$-Komponenten der Vektoren separat.

Zuerst die $x-$Richtung:

$$T_{2x}+T_{3x}=0$$ 

Die Komponenten können durch den Betrag und Winkel ausgedrückt werden: 

$$-T_2\sin\alpha+T_3\sin\beta=0\tag{4.1}$$

Für die $y-$Richtung:

$$-m\,g+T_2\cos\alpha+T_2\cos\beta=0\tag{4.2}$$

Merke: $\sin 30^\circ=\cos 60^\circ=\frac{1}{2}$ und $\cos 30^\circ=\sin 60^\circ=\frac{\sqrt{3}}{2}$.


Gleichung (4.1) kann umgeformt werden nach 

$$T_2=\sqrt{3}T_3$$ 

und in Gleichung (4.2) eingesetzt und aufgelöst nach $T_3$:

$$T_3=\frac{m\,g}{2}$$
und damit
$$T_2=\frac{\sqrt{3}}{2}m\,g$$


Es geht noch einfacher. Da in Ruhelage die Vektoren $\vec{T}_2$ und $\vec{T}_3$ identisch mit $\vec{F}_1$ sein muss, kann man wegen der Winkel von $\alpha+\beta=90^\circ$ das pythagoraische Dreieick benutzen:

$$T_2^2+T_3^2=m\,g$$

#### 3. Newtonsche Gesetz

<b>Wenn ein Körper mit einer Kraft auf einen zweiten Körper wirkt, wirkt der zweite Körper mit der gleiche großen Kraft aber entgegengesetzen Richtung. Jede Kraft hat eine Gegenkraft: actio = -reactio.</b>

*Beispiel:*

Hier betrachten wir die Kraft und deren Übertragung auf zwei Körper. $\vec{F}$ wirkt auf Körper 1, und der wieder auf Körper 2. Die Massen sind $m_1=5\,$kg und $m_2=15\,$kg, und die Kraft $F=20\,$N.

<img src="figures/4.1_6.png" width=250px>

Beide Körper haben bei Kontakt die gleiche Beschleuningung, also ist wegen

$$F=m\,a=(m_1+m_2)\,a$$

die Beschleuningung $a=1\,$ms$^{-2}$. 

<img src="figures/4.1_7.png" width=180px>

Betrachten wir nun Masse $m_2$. Auf diese wirkt die Kontaktkraft von Masse $m_1$ auf $m_2$, $F_{12}$. Diese muss 

$$F_{12}=m_2\,a=15\,\mathrm{N}$$

sein. Die Krafteinwirking auf $m_1$ ist hier dargestellt:

<img src="figures/4.1_8.png" width=200px>

Die Kraft $F$ wirkt nach rechts, und die Kontaktkraft, $F_{21}$, von Masse $m_2$ auf Masse $m_1$ nach links. Daher

$$F+F_{21}=m_1\,a=5\,\mathrm{N}$$

$$20\,\mathrm{N}+F_{21}=5\,\mathrm{N}$$

$F_{21}=-15\,$N. Das zeigt schön, dass actio = - reactio ist, also $F_{12}=-F_{21}$.



*Konsequenzen des 3. Newtonschen Gesetzes:*

Gleiches gilt für alle Kräfte, selbst wenn ein Apfel vom Baum fällt:


<img src="figures/4.1_9.png" width=250px>

Die Masse des Apfels sei $m_{\mathrm{Apfel}}=0.5\,$kg, die Masse der Erde $m_{\mathrm{Erde}}=6\cdot 10^{26}\,$kg und der Apfel fällt aus einer Höhe von $2\,$m.

Auf den Apfel wirkt die Gravitationskraft der Erde $F=m\,g$. Gleichfalls nach dem 3. Newtonschen Gesetz wirkt auf die Erde die Gegenkraft $-F$.
Wie weit bewegt sich die Erde auf den Apfel zu, wenn der Apfel von der Höhe $h$ fällt?


$$F=m\,g=0.5\,\mathrm{kg}\,10\,\mathrm{m}\mathrm{s}^{-2}=5\,\mathrm{N}$$

Die Dauer des Falls ergibt sich aus

$$h=\frac{1}{2}\,g t^2\quad .$$

Nach $t$ aufgelöst ergibt sich eine Fallzeit $\approx 0.3\,$s.

Die Erde erfährt als Konsquenz des 3. Newtonschen Gesetzes die gleiche Kraft aber in der entgegengesetzen Richtung. 

$$F_{\mathrm{Erde}}=-5\,\mathrm{N}=m_{\mathrm{Erde}}\,a_{\mathrm{Erde}}$$

Die Beschleunigung der Erde $a_\mathrm{Erde}=8.3\cdot 10^{-27}\,$m/s$^2$. Während der Fallzeit von $0.3\,$s bewegt sich die Erde um $3.8\cdot 10^{-28}\,$m auf den Apfel zu.
