# 5. Trägheitskräfte
## 5.1 Inertialsystem

Die Position des Massepunktes $m$ kann im Koordinatensystem $\Sigma$ mit den Koordinaten $(x,y,z)$ oder im Koordinatensystem $\sigma'$ mit den Koordinaten $(x',y',z')$ beschrieben werden.

<img src="figures/4.3_5.png" width=250px>

Beide Koordinatensystem bewegen sich mit der Geschwindigkeit $\vec{v}_{rel}$ voneinander weg.

<b>Alle nichtbeschleunigten Koordinatensysteme heißen Inertialsysteme. Hier gelten die Newtonsche Axiome.</b> 

## 5.2 Geradlinig beschleunigte Bezugssysteme

d.h. $\Sigma'$ wird gegenüber $\sigma$ beschleunigt ⇒ $\vec{v}_{rel}\ne \mathrm{konst.}$

<img src="figures/4.3_6.png" width=250px>

Die Beschleunigung ist nur entlang der $x$-Richtung.

$$x=x'+s\quad y=y'\quad z=z'$$

$$\ddot{x}=\ddot{x'}+\ddot{s}$$

$$\ddot{\vec{r}}=\ddot{\vec{r'}}+\ddot{\vec{s}}$$

$$\ddot{\vec{a}}=\ddot{\vec{a'}}+\ddot{\vec{s}}$$

$$m\ddot{\vec{a}}=m\ddot{\vec{a'}}+m\ddot{\vec{s}}$$

$$\vec{F}=\vec{F}'+m\,\ddot{\vec{s}}$$

$$\vec{F'}=\vec{F}\underbrace{-m\,\ddot{\vec{s}}}_{\vec{F}_T~\mathrm{Trägheitskraft}}$$

$$\vec{F}_T=-m \ddot{\vec{s}}$$

Im beschleunigten Koordinatensystem wirkt die Trägheitskraft in entgegengesetzer Richtung zum Beschleunigungsvektor. Damit erklärt sich, warum wir im freien Fall keine Kraft spüren: die Gravitationskraft und die Trägheitskraft heben sich genau auf.


## 5.3 Rotierende Bezugssyteme

### 5.3.1 Zentrifugalkraft

<img src="figures/5.3_1.png" width=250px>

Die Zentrifugalkraft stellt die Trägheitskraft in einem rotierenden Bezugsystem mit $\vec{v}'=\mathrm{konst.}$ dar.

Dann wirken im rotierenden Bezugssystem die Kräfte

$$\vec{F}_{r} + \vec{F}_{Zf} = 0$$

und damit ist 

\begin{eqnarray}
\vec{F}_{Zf}&=&-\vec{F}_r\\
&=&-\left(-m\frac{v^2}{r}\vec{e}_r\right)
\end{eqnarray}

Also ist die Zentrifugalkraft:
$$\vec{F}_{Zf}=m\,\frac{v^2}{r}\vec{e}_r$$

### 5.3.2 Corioliskraft

Wenn die Relativbewegung im rotierenden Koordinatensystem $\vec{v}\ne 0$, dann wirkt die Corioliskraft.

$$\vec{F}_C=2\,m\left(\vec{v}'\times\vec{\omega}\right)$$

$\vec{\omega}$ ist der Vektor der Winkelgeschwindigkeit. Der Betrag von $\vec{\omega}$ gibt die Winkelgeschwindigkeit an und die Richung die Orientierung der Drehaches. Dies ist bei der Bestimmung der Richtung der Corioliskraft zu berücksichtigen.

Wir betrachten die Corioliskraft zuerst auf einer Scheibe. In der Animation unten schauen wir auf eine lineare Trajektorie von einem Inertialsystem auf eine rotierende Scheibe. Die untere Ansicht ist von einem Beobachter, der auf der Scheibe z.B. an Ort, der mit einem roten Punkt markiert ist. Dort ist die Trajektorie gekrümmt und für den Betrachter wirkt die Corioliskraft auf das Objekt.
