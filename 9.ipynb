# 9. Ruhende Flüssigkeiten und Gase

### 9.1.1 Freie Verschiebbarkeit

* Ideale Fluide (Gase und Flüssigkeiten)
  * keine Reibungseffekte
  * keine Oberflächeneffekte bei Flüssigkeiten
* Unter dem Einfluß der Schwerkraft
  => ebene Oberfläche
  
* freie Verschiebbarkeit der Moleküle (z.B. Wasser)

Beispiel für eine nichtideale Flüssigkeit, das [längste laufende Experiment](https://www.nature.com/news/world-s-slowest-moving-drop-caught-on-camera-at-last-1.13418).

### 9.1.2 Dichte

Die Dichte $\rho$ eines Körpers ist definiert als seine Masse pro Volumen:

$$\rho=\frac{m}{V}\quad ,$$ 

die SI-Einheit $[\rho]=$kg/m$^3$.

Beispiele der Dichte von festen, flüssigen und gasförmigen Stoffen.

|Stoff|Dichte in kg/m$^3$|
|---|---|
| Eisen | $7,8\cdot 10^3$ |
| Blei | $11,3\cdot 10^3$|
| Holz | $0,3-0,9 \cdot 10^3$|
| Eis | $0,917\cdot 10^3$|
| Wasser (4$^\circ$C)|$0,917\cdot 10^3$|
| Quecksilber | $13,6\cdot 10^3$|

Die Dichte in Flüssigkeiten ist nahezu konstant, in Gasen kann sie sich aber leicht durch Krafteinwirkung ändern. Wie nennen dieses Verhalten kompressibel. Flüssigkeiten sind inkompressibel.

### 9.1.3 Druck in Flüssigkeiten

Der Druck $p$ ist die Kraft pro Flächeneinheit. Wenn eine homogene Kraft $F$ senkrecht auf die Fläche $A$ wirkt ist der Druck. 

$$p=\frac{F}{A}$$


$$\left[p\right]=\frac{\mathrm{N}}{\mathrm{m}^2}=\mathrm{Pa}$$

Der atmosphärische Normaldruck ist $1013\,\mathrm{hPa}=1\,\mathrm{atm}=760\,\mathrm{Torr}$.

### 9.1.4 Gesetz von Pascal

Der Druck in einem ruhenden Fluid (ohne Einbeziehung der Schwerkraft) ist in allen Punkten und in allen Richtungen gleich.

Auf eine Gefässwand wirkt eine Kraft $\vec{F}$, die Kraft der Gefäßwand auf die Flüssigkeit ist $-\vec{F}$.

<img src="figures\9.1_1.png" width=200px>

Allgemein kann man den Druck als die Grösse pro Flächenelement $\mathrm{d}\vec{A}$ definieren, dass eine Kraft $\mathrm{d}\vec{F}$ senkrecht zur Fläche erzeugt.

$$\mathrm{d}\vec{F}=-p{\mathrm{d}\vec{A}}=-p \,\vec{n}\,\mathrm{d}A$$

Die auf den Druck des Fluids zurückzuführende Kraft ist immer senkrecht zu jeder Fläche und wirkt auf Fläche mit der das Fluid Kontakt hat.

### 9.1.5 Schweredruck 

Wir berechnen nun die durch die Schwerkraft ausgeübte Druckänderung in einer Flüssigkeit. Die Dichte kann in guter Näherung als konstant angenommen werden und wir erhalten als Schweredruck 

<img src="figures\9.1_2.png" width=200px>

Schweredruck:

$$p_S=\frac{F}{A}=\frac{mg}{A}=\frac{\rho_{Fl} A g h}{A}=\rho_{Fl} g h$$

Hydrostatischer Druck:

$$p_H=p_S+P_{\mathrm{Luft}}=p_{ges}(h)$$

### 9.1.6 U-Rohr Manometer

Das wissen, dass die Druckdifferenz proportional mit der Tiefe $h$ in einer Flüssigkeit proportional kann zum Messen von unbekannten Drücken ausgenützt werden. Die unten stehende Abbildung zeigt ein einfaches Druckmessgerät, das offene Manometer, das wegen seiner Form auch U-Rohr-Manometer genannt wird.

### 9.1.7 Auftrieb

* Archimedisches Prinzip

<img src="figures\9.1_3.png" width=220px>

$$\vec{F}_1=-m_{Fl}^{(1)}\,g\,\vec{e}_x$$

$$\vec{F}_2=+m_{Fl}^{(2)}\,g\,\vec{e}_x$$

$$F_{A}=\vec{F}_1+\vec{F}_2=g \, \rho_{Fl}\,A\,(h_2-h_1)\,\vec{e}_x$$

Das Volumen des Körpers ist $A(h_2-h_1)=V_{\mathrm{Körper}}$.

Und die Masse der verdrängten Flüssigkeit ist $m_{Fl}=\rho_{Fl}\,V_{\mathrm{Körper}}$.

$$\vec{F}_{A}=\rho_{Fl}\, g\,  V_{\mathrm{Körper}}\,\vec{e}_x$$.

*Das Archimedische Prinzip* Die Auftriebskraft, die ein Körper, der in ein Fluid eingetaucht ist, erfährt, ist gleich der Gewichtskraft des durch diesen Körper verdrängten Fluids.


Das scheinbare Gewicht $F'_G$ eines in ein Fluid eintauchenden Körpers ist die Differenz zwischen dem Gewicht $|\vec{F}_G|$ und der Auftriebskraft $|\vec{F}_A|$:

$$F'_G=|\vec{F}_G|-|\vec{F}_A|$$


Daraus lassen sich für $F'_G$ folgende Fälle ableiten:

* $F'_G>0$ sinkt der Körper
* $F'_G=0$ schwebt der Körper
* $F'_G<0$ schwimmt der Körper

## 9.2 Erscheinungen an Grenzflächen
### 9.2.1 Oberflächenspannung

Kohäsionskräfte: zwischen den Molekülen im Inneren der Flüssigkeit

$$ \vec{F}_{ij}=0$$

<img src="figures/9.2_1.png" width=350px>

An Rändern der Flüssigkeit, also zu festen und gasförmigen Stoffen, oder anderen Flüssigkeiten wirken die Adhäsionkräfte. Diese können stärker oder schwächer als die Kohäsionskraft sein. Bei Gasen ist die Adhäsion wesentlich geringer und auf die Flüssigkeitsmoleküle an der Grenzfläche flüssig-gasförmig wirkt eine resultierende Kraft in die Flüssigkeit hinein.



$$ \vec{F}_{ij}=\vec{F}_{res}\ne0$$

#### Oberflächenspannung
* Zur Vergrößerung von Oberflächen einer Flüssigkeit müssen Moleküle an die Oberfläche gebracht werden => Arbeit gegen $\vec{F}_{res}$.
* Vergrößerung der Oberflächenenergie:

  $$\sigma=\frac{\Delta W}{\Delta A}$$
  
  $\sigma$: spezifische Oberflächenenergie
  
  $\Delta W$: Arbeit
  
  $\Delta A$: Änderung der Oberfläche
  
  
* Bestimmung von $\sigma$ mittels Bügelmethode


<img src="figures/9.2_2.png" width=300px>  

$$\Delta W=F\,\Delta s$$

$$\sigma=\frac{F\,\Delta s}{2\,\Delta s\,l}=\frac{F}{2\,l}$$

* Druck in Seifenblasen

<img src="figures/9.2_3.png" width=350px>  

 $p$ ist Druck im Inneren der Blase, $p_0$ der äußerer Luftdruck, $A$ die Oberfläche der Blase. 
 
 Die Blase versucht in den Zustand geringster Oberflächenenergie zu gelangen $\rightarrow$ Verkleinerung der Oberfläche. 
 
 Im Inneren herrscht ein Überdruck $\Delta p_{SB}=p-p_0$. Dieser führt zu einer radial nach außen gerichteten Kraft 
 
 $$F_p=\Delta p_{SB}\,A$$
 
 Wir wollen nun diese Durckdifferenz bestimmen:
 
 Der Energiegewinn $\Delta W$ beim Verkleinern der Oberfläche ist
 
 \begin{eqnarray}
 \Delta W=\sigma \, \Delta A&=&2\sigma\,4\pi\left[\left(r+\Delta r\right)^2-r^2\right]\\
 &=&2\sigma\,4\pi\left[2\,r\,\Delta r+\Delta r^2\right]\\
 &\approx&4\sigma\,4\pi(\,r\,\Delta r)\quad ,
 \end{eqnarray}
 da $\Delta r^2$ vernachlässigt werden kann für $\Delta r\rightarrow 0$. Der Faktor 2 kommt von den beiden Oberflächen der Blasen.
 
 Die Arbeit zu leisten gegen $\Delta p$
 
 $$\Delta W_A=F_P\,\Delta r=\Delta p_{SB}\, \underbrace{4\,\pi\,r^2}_A\,\Delta r$$
 
 Gleichgewicht herrscht wenn 
 
 \begin{eqnarray}
 \Delta W &=& \Delta W_A\\
 4\sigma &=& \Delta p_{SB} \, r\\
 \Delta p_{SB} &=& \frac{4\sigma}{r}
 \end{eqnarray}
 
 Die Druckdifferenz $\Delta p$ wird grösser, je kleiner die Seifenblase wird.
 
 
 Für Blasen in Flüssigkeit gilt sehr ähnliche Herleitung, nur dass dann nur *eine* Oberfläche berücksichtigt werden muss. Wir erhalten somit
 
 $$\Delta p_B = \frac{2\sigma}{r}\quad .$$
 
 ### 9.2.2 Grenzflächenspannung & Benetzung
 
* Benetzung

 <img src="figures/9.2_4.png" width=350px>  
 
 Wir sprechen von Benetzung wenn die Adhäsionskräfte größer als die Kohäsionskräfte sind. Es werden die Kräfte für die drei Phasen: fest (3), flüssig (2) und gasförmig (1) unterschieden. Wenn der Tropfen eine feste Position mit dem Kontaktwinkel $\alpha$ einnimmt, müssen sich die Krâfte zu null addieren. Das Kräftegleichgewicht ergibt

 $$\sigma_{13}=\sigma_{23} + \sigma_{12}\cos\alpha$$
 
 Wenn der Kontaktwinkel $\alpha<=\pi/2$ benetzt die Flüssigkeit die Oberfläche.
 
 <img src="figures/9.2_5.png" width=270px> 
 
 Wenn der Kontaktwinkel $\alpha>\pi/2$ findet keine Benetzung statt. Die Flüssigkeit zieht sich zu einem Tropfen zusammen. Die Kohäsionsskräfte sind also größer als die Adhäsionskräfte.
 
 Im Falle von Wasser spricht man von hydrophil, wenn die Oberfläche benetzt wird, und hydrophob wenn sie nicht benetzt wird.
 
### 9.2.3 Kapillarität
 
 In Kapillaren (dünne Röhrchen mit Radius $r$ ), die sich benetzten, steigt der Flüssigkeitsspiegel. Das nennt man Kapillaraszension. In nicht benetzenden Kapillaren (z.B. Glas und Quecksilber), fällt der Flüssigkeitsspiegel, die Kapillardepression.
 
 <img src="figures/9.2_6.png" width=500px>  
 
 Die Oberflächenspannung und die Gewichtskraft gleichen sich in einer solchen Kapillare aus. Die von der Oberflächenspannung herrührende Kraft $F_\sigma$ ist gleich der Gewichtskraft $F_g$, also $F_\sigma=F_g$. 
 
 $$F_\sigma=\sigma\,2\pi\,r$$
 
 $$F_g=m\,g\,h=\rho_{Fl}\pi\,r^2\,h$$
 
 Umstellen nach $h$ ergibt eine Steighöhe von
 
 $$h=\frac{2\sigma_{12}\cos\alpha}{\rho_{Fl} g r}$$
 
 Hier haben wir noch die Beziehung zwischen Oberflächenspannung $\sigma_{12}$ und Kontaktwinkel benutzt.
 
 Die Steighöhe kann positiv und negativ sein, abhängig davon ob die Flüssigkeit die Oberfläche benetzt oder nicht. Die Steighöhe erhöht sich für kleine Kapillare. Dies ist zum Beispiel wichtig für Pflanzen und Bäume, die Kapillare für den Flüssigkeitstransport benutzen.
