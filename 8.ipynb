* Gesamtenergie der Bewegung?

  Benutzung von Freiheitsgraden
  * 3 Freiheitsgrade der Translation
  
    Kinetische Energie der Translation
  
    $$ E_{kin}=\frac{1}{2}m v^2=\frac{1}{2}m\left( v_x^2+v_y^2+v_z^2\right)$$
  
  * 3 Freiheitgrade der Rotation (um die möglichen Rotationsachsen)

## 8.2 Drehimpuls und Rotationsenergie

<img src="figures/8.2_1.png" width=280px>

Der Drehimpuls eines Körpers mit der Geschwindigkeit $\vec{v}$ und Masse $m$ (Impulse $\vec{p}$ ist gegeben durch

$$\vec{L}_A=\vec{r}\times(m\,\vec{v})=\vec{r}\times\vec{p}$$

Hier $\vec{r}$ ist der Ortsvektor zum Bezugspunkt $A$ ist. Der Drehimpuls $\vec{L}$ ist immer abhängig vom Bezugspunkt, und daher schreiben wir $\vec{L}_A$.

* Betrag von $\vec{L}$

  $$\left|\vec{L}\right|=\left|\vec{r}\right|\,\left|\vec{p}\right|\,\sin\left(\vec{r},\vec{p}\right)=\left|\vec{r}\right|\,\left|\vec{p}\right|\,\sin \gamma$$
  
  * $\left|\vec{L}\right|$ wird maximal wenn $\vec{r}\perp\vec{v}$
  
  * $\left|\vec{L}\right|$ wird minimal wenn $\gamma=0^\circ$ oder $\gamma=90^\circ$, also $\vec{r}\parallel\vec{v}$
  
  * Richtung von $\vec{L}$ mit der Rechten-Hand-Regel
  
  
* $\vec{L}_Z$ einer Punktmasse auf einer Kreisbahn


  <img src="figures/8.2_2.png" width=240px>
  
  Wir wählen als Bezugspunkt den Kreismittelpunkt $Z$ aus. Mit Hilfe der Rechten-Hand-Regel sehen Sie, dass der Drehimpuls aus der Bildebene zu Ihnen zeigt. Damit ist die Richtung identisch mit der Richtung der Drehachse $\vec{\omega}$. Der Betrag läßt sich leicht bestimmen, da $\vec{r}_Z\perp\vec{v}$ ist.

  \begin{eqnarray}
  \left|\vec{L}_Z\right|&=&m\,\left|\vec{r}_Z\times\vec{v}\right|\\
  &=&m\,r_Z\,v=m\,{r_Z}^2\,\omega
  \end{eqnarray}
  
  Da wir die Richtung von $\vec{L}_Z$ kennen können wir gleich
  
  $$\vec{L}_Z=m\,{r_Z}^2\,\vec{\omega}$$
  
  schreiben.
  
  $J_{A,\vec{\omega}}=m\,r^2$ ist das Massenträgheitsmoment bezüglch einer Achse $\vec{\omega}$ durch den Punkt $A$, dann ist der Drehimpuls
  
  $$\vec{L}=J_{A,\vec{\omega}}\,\vec{\omega}$$
  
  
  Jetzt beschreiben wird die Rotationsenergie des einen Massenpunktes $m$ durch die kinetische Energie des Systems:
  
  $$E_{kin}=\frac{1}{2}m\,v^2=\frac{1}{2}m\,r_Z^2\,\omega^2$$
  
  Hier ist die Bewegung eine reine Rotation, d.h. die kinetische Energie besteht nur aus der Rotationsenergie
  
  $$E_{rot}=E_{kin}$$
  
  und die Konstante in der Rotationsenergie ist $J_{Z,\vec{\omega}}=m\,r_Z^2$.
  
  $$E_{rot}=\frac{1}{2}J_{Z,\vec{\omega}}\,\omega^2$$
  
  $J_{Z,\vec{\omega}}$ ist das Massenträgheitsmoment der Rotationsbewegung um eine Achse die durch den Punkt $Z$ in Richtung $\vec{\omega}$ geht.

## 8.3 Massenträgheitsmoment 

Bestimmung des Massenträgheitsmomentes für einen ausgedehnten Körper

<img src="figures\8.2_3.png" width=300px>

Dazu brauchen wir eine Drehachse $\vec{\omega}$ und einen Drehpunkt $A$ durch den die Drehachse geht. Dann zerlegen wir den Körper in Massenelemente $\Delta m_i$

$$ m=\sum_i^N \Delta m_i$$

Dann ist das Massenträgheitsmoment

$$J_{A,\vec{\omega}}=\sum_i^N\,\Delta m_i\,{r_{A_i}}^2\quad .$$

Für einen kontinuierlichen Körper lassen wir die Massenelemente $\Delta m_i\rightarrow 0$ gehen und benutzen infintisimale kleine Massen).

$$\rightarrow\sum\rightarrow\int$$

$$J_{A,\vec{\omega}}=\int {r_A}^2 \, \mathrm{d}m$$


#### Beispiel für einen Zylinder

mit konstanter Dichte, der um die Symmetrieachse $\vec{\omega}=\omega\,\vec{e}_z$ durch das Zentrum $S: x=y=z=0$ rotiert:

<img src="figures\8.3_2.png" width=200px>

$$J_{S,z}=\int r^2 \,\mathrm{d}m\quad .$$

Bei konstanter Dichte ist $\mathrm{d}m=\rho\,\mathrm{d}V$ und

$$J_{S,z}=\rho\int r^2 \,\mathrm{d}V\quad .$$

Für die Integration über $\mathrm{d}V$ muß ein geeignetes Volumenelement genommen werden. Hier bieten sich zylindrische Koordinaten an. Diese kann man leicht verstehen:

<img src="figures\8.3_3.png" width=250px>

Das Flächenelement $\mathrm{d}A$ eines Querschnitts des Zylinders ist

$$\mathrm{d}A=\underbrace{r\,\mathrm{d}\varphi}_{\mathrm{Bogenlänge}}\,\mathrm{d}r$$

Das Volumen beruücksichtigt dann nur noch die Tiefe in $z$-Richtung:

$$\mathrm{d}V=r\,\mathrm{d}\varphi\,\mathrm{d}r\,\mathrm{d}z$$

$$J_{S,z}=\rho\int r^2 \,\mathrm{d}V\quad=
\rho\int_0^h\int_0^R\int_0^{2\pi}r^2\,r\,\mathrm{d}\varphi\,\mathrm{d}r\,\mathrm{d}z$$

Integration über $\mathrm{d}\varphi$ ergibt $2\pi$

$$J_{S,z}=2\pi\,\rho\int_0^h\int_0^R r^3\mathrm{d}r\,\mathrm{d}z$$

Integration über $\mathrm{d}z$ ergibt $h$

$$J_{S,z}=2\pi\,\rho h \int_0^R r^3\mathrm{d}r$$

Und zum Schluss über $r$

$$J_{S,z}=2\pi\,\rho h \left[\frac{r^4}{4}\right]_0^R=\underbrace{\rho \pi R^2 h}_{\rho\,V=m} \,\frac{1}{2}R^2=\frac{1}{2}m R^2$$



#### Rollbewegung
Eine Rollbewegung besteht aus der Translation des Massenzentrums und einer Rotation um eine Achse durch den Schwerpunkt des Rollkörpers. Beim idealen Rollen ohne zu gleiten sind beide Geschwindigkeiten durch die Beziehung $v=\omega\,R$ gegeben.
<img src="figures\8.3_4.png" width=350px>


#### Geschwindigkeit eines rollenden Zylinders
Wir können die Geschwindigkeit eines Zylinders, der eine schiefe Ebene hinab rollt leicht mit dem Energieerhaltungssatz bestimmen. Dazu nehmen wir an, dass keine Reibungsarbeit geleistet wird (der Kontaktpunkt zwischen Zylinder und Ebene ruht).

<img src="figures\8.3_5.png" width=350px>


Zu Beginnn sei der Zylinder in Ruhe, dann ist 

$$E_{pot}=m g h$$

Danach wird er schneller. Wir haben zwei kinetische Energien durch die Translation und die Rotation:

$$E_{kin}=\frac{1}{2} m v^2+\frac{1}{2}J_S\,\omega^2$$

Nachdem der Zylinder die Höhe $h$ heruntergerollt ist, haben wir die potentielle Energie in kinetische Energie umgewandelt, also ist die Geschwindigkeit

$$v=\sqrt{\frac{2 g h}{1+\frac{J_s}{m R^2}}}$$

Im Falle eines Zylinders mit $J_S=\frac{1}{2}mR^2$

erhalten wir eine Geschwindigkeit von 

$$v=\sqrt{\frac{4}{3}g\,h}<\sqrt{2 g h}$$

Die Geschwindigkeit ist kleiner als die Geschwindigkeit eines Massenpunktes, der eine Höhe $h$ frei fällt, nämlich $v=\sqrt{2 g h}$. Warum?


## 8.4 Satz von Steiner 

<img src="figures\8.4_1.jpg" width=250px>

Das Trägheitsmoment $J_A$ eines Körpers um die Achse $A$ setzt sich additiv aus dem Trägheitsmoment $J_S$ bezüglich des Schwerpunktes und dem Trägheitsmoment $m\,a^2$. Dabei bezeichnet $a$ den Abstand der beiden parallelen Achsen und $m$ die gesamte Masse des Körpers. <b>Wichtig</b> Beide Achsen müssen parallel sein.





## 8.5 Dynamisches Grundgesetz der Rotation

* Bei der Translation führt eine Kraft zu einer Impulsänderung

  $$ \vec{F}=\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}$$

* Wann ändert sich der Drehimpuls?
  Dazu bestimmmen wir die zeitliche Änderung von $\vec{L}$:
  
  \begin{eqnarray}
  \frac{\mathrm{d}\vec{L}}{\mathrm{d}t}&=&\frac{\mathrm{d}\,(\vec{r}\times\vec{p})}{\mathrm{d}t}\\
  &=&\underbrace{\frac{\mathrm{d}\vec{r}}{\mathrm{d}t}\times\vec{p}}_{\vec{v}\times\vec{p}=0}+\vec{r}\times\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}\\
  \frac{\mathrm{d}\vec{L}}{\mathrm{d}t}&=&\vec{r}\times\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}=\vec{r}\times\vec{F}
  \end{eqnarray}

* Die zeitliche Änderung des Drehimpulses wird Drehmoment $\vec{M}$ genannt

  $$\vec{M}=\vec{r}\times\vec{F}$$
  
  Der Drehimpuls ist auf einen Körper ist
  
  $$\vec{L}=J_A\,\vec{\omega}$$
  
  und damit ist 
  
  $$\vec{M}=\frac{\mathrm{d}\vec{L}}{\mathrm{d}t}=
  \frac{\mathrm{d}J_A}{\mathrm{d}t}\vec{\omega}+J_A\,\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}$$

  Die Winkelbeschleunigung ist 
  
  $$\vec{\alpha}=\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}$$
  
  und für den Fall eines konstanten Trägheitsmoments erhalten wir das *Grundgesetz der Rotation*
  
  $$\vec{M}=J_A\,\vec{\alpha}$$


## 8.6 Drehimpulserhaltungssatz

* abgeschlossenes physikalisches System

  $$\vec{M}=\sum_{i=1}^N\vec{M}_i$$
  
  Es wirkt keine äußere Kraft an, deshalb ist $\vec{M}=0$ und da $\vec{M}=\frac{\mathrm{d}\vec{L}}{\mathrm{d}t}$ ist $\vec{L}=\mathrm{konst}$.
  
* für mehrere Körper im System

  $$\vec{L}=\sum_{i=1}^N\vec{L}_i$$
  
  
* Beispiel 1: $\vec{L}=0$

  für zwei Körper:
  
  $$\vec{L}=J_1\vec{\omega}_1+J_2\vec{\omega}_2=0$$
  
  $$\vec{\omega}_2=-\frac{J_1}{J_2} \vec{\omega}_1$$
  
  
* Beispiel 2: $\vec{L}=\mathrm{konst.}\ne0$

  Ausgangszustand
  $$\vec{L}= J_1\vec{\omega}_1$$
  
  Endzustand
  $$\vec{L}= J_2\vec{\omega}_2$$
  
  $$\vec{\omega_2}=\frac{J_1}{J_2}\vec{\omega}_1$$

## 8.7 Kreisel: Freie Achsen

* Bisher: Rotation um raumfeste Achsen

* Jetzt: Bewegung starrer Körper ohne Fixierung der Rotationsachse $\rightarrow$ Kreisel.

Wir betrachten nur Rotationen durch den Schwerpunkt. Die Symmetrieachsen des Objektes nennen wir $\vec{a}$, $\vec{b}$ und $\vec{c}$.

Allgemein hat der Kreisel für jede Symmetrieachse ein Trägheitsmoment.

Die Achse mit dem grössten Trägheitsmoment ist $\vec{c}$. Die mit dem niedrigsten $\vec{a}$: $J_a\le J_b\le J_c$

Symmetrische Kreisel sind Kreisel in denen mindestens 2 Trägheitsmomente gleich groß sind.

<img src="figures/8.7_1.gif" width=500px>

* prolater Kreisel: $J_a<J_b=J_c$

* oblater Kreisel: $J_a=J_b<J_c$

Die Rotation parallel zu $\vec{a}$ und $\vec{c}$ stabil sind.

## 8.8 Kreisel: Präzession

Wir betrachten ein Rad das auf einer Achse in Richtung der $x$-Achse gelagert ist.

<img src="figures/8.8_1.png" width=200px>

Das Rad steht still und wird nun durch ein Drehmoment $\vec{M}$ das für eine kurze Zeit $\Delta t$ wirkt ausgelenkt. Wir stellen uns den Versuch im Weltraum ohne äußere Kräfte vor. Das Drehmoment kann bespielsweise durch ein Kräftepaar erzeugt werden, die beide mit gleicher Kraft aber in entgegengesetzer Richtung am Schwerpunkt des Rades angreifen.

<img src="figures/8.8_2.png" width=500px>

Das Rad erhält einen Drehimpuls da

$$\vec{M}=\frac{\mathrm{d}\vec{L}}{\mathrm{d}t}\approx\frac{\Delta \vec{L}}{\Delta t}$$

und dreht sich folglich um die $z$-Achse mit Drehimpuls $L_z\vec{e}_z$.

#### Mit Eigendrehimpuls

Jetzt wird dem Rad ein Eigendrehimpuls gegeben, also eine Drehimpuls um den Schwerpunkt um die Symmetrieachse. Das Rad dreht sich um die Achse mit Winkelgeschwindigkeit $\omega$. Wiederum wirkt kurzfristig ein Drehmoment am Rad durch das Kräftepaar $\vec{F}$. Jetzt stellst sich die Frage, wie sich das Rad nach der Zeit $\Delta \tau$ bewegt?
Die Antwort ist gar nicht so einfach. Denn, nachdem das Drehmoment gewirkt hat muss der Drehimpuls $\vec{L}=\mathrm{konst.}$ sein. D.h. das Rad kann sich nicht einfach wie oben um die $z$-Achse drehen. Dann würde sich ja das Drehmoment kontinuierlich seine Richtung ändern. Die Antwort ist in der Abbildung beschrieben.

<img src="figures/8.8_3.png" width=500px>

Durch das Wirken des Drehimpulses $\vec{M}$ ändert sich der Drehimpuls um den Wert $\vec{\Delta L}$, die Drehachse wird also verändert. Nach der Zeit $\Delta t$ bleibt die neue Drehachse erhalten.


Wir wollen das am zweiten Beispiel nochmal verdeutlichen. Das Rad dreht sich mit dem gleichen Eigendrehimpuls $\vec{L}$, jetzt wird aber ein Drehmoment kurzfzeitig in Richtung der negative $x$-Achse ausgeübt, $\vec{M} wie in der untenstehenden Abbildung links angedeutet.

<img src="figures/8.8_4.png" width=500px>

Das Ergebnis ist eine Änderung des Eigendrehimpulses $\vec{L}$ in Richtung des Drehmomentes, das Rad rotiert kurz um die $z$-Achse und stellt die Drehimpulsachse auf die neue $\vec{L}$ Richtung ein.

Die Änderung der Richtung des Drehimpulses eines Kreisels wird *Präzession* genannt. Wir können die Änderung der Ausrichtung des Kreisels mit einem einfachen Experiment auch quantitativ beschreiben. Dazu wird der Rad-Kreisel zum Rotieren gebracht und drehbar an einem Seil aufgehängt. An dem Schwerpunkt des Kreisels wirkt die Gewichtskraft $\vec{F_g}$, das ein Drehmoment erzeugt.

<img src="figures/8.8_5.png" width=200px>

Das Drehmoment ist 

$$\vec{M}=\vec{PQ}\times \vec{F_g}$$

Das Drehmoment $\vec{M}$ zeigt in die Bildebende hinein. Der Winkel zwischen $\vec{PQ}$ und $\vec{F_g}$ ist $90^\circ$ und damit ist der Betrag von

$$\left|\vec{M}\right|=r\,m\,g\quad .$$

Das Drehmoment ändert den Drehimpuls. Der Drehimpuls folgt dem Drehmoment. Der Kreisel rotiert um die $z$-Achse gegen den Urzeigersinn mit einer konstanten Winkelgeschwindigkeit der Präzession. 


Die Winkelgeschwindigkeit der Präzession $\omega_{pr}$ wird ohne Herleitung hier angegegeben:

$$\omega_{pr}=\frac{M}{L_s}\quad,$$

wobei $L_s$ der Betrag des Eigendrehimpuls des Kreisels ist und $M$ der Betrag des Drehmomentes. Diese Formel gilt nur wenn der Eigendrehimpuls sehr viel größer als der Drehimpuls durch die Präzession ist.

Das Drehmoment ist

$$M=r\,m\,g$$

und der Drehimpuls ist

$$L=I_{\mathrm{Rad}}\,\omega\quad.$$

Für ein Rad, bei dem die ganze Masse auf einem Ring mit Radius $R$ verteilt ist, ist $I_{\mathrm{Rad}}=m\,R^2$. $\omega$ ist die Winkelgeschwindigkeit des Eigendrehimpulses (des Spins).

$$\omega_{pr}=\frac{r\,m\,g}{m\,R^2\,\omega}=\frac{r\,g}{R^2\,\omega}$$

### 8.10.1 Gravitationsgesetz

<img src="figures/8.10_1.png" width=350px>

Massen ziehen sich an. Wir haben bisher die Gravitationskraft $\vec{F}_g=m\,\vec{g}$ benutzt und $\vec{g}$ die "konstante" Erdbeschleunigung benutzt. Das ist eine Vereinfachung des allgemeinen Gravitationsgesetzes das die Anziehung zweier Massen $m_1$ und $m_2$ beschreibt:

$$\vec{F}_G=-\gamma\,\frac{m_1\,m_2}{r^2}\,\vec{e}_r\quad .\tag{8.1}$$

Hier ist $r$ der Abstand zwischen den Massen mit Abstandsvektor $\vec{r}=r\,\vec{e}_r$ und $\gamma$ ist die allgemeine Gravitationskonstante:

$$\gamma=6.67408\pm0.00031\cdot 10^{-11}\mathrm{m}^3\mathrm{kg}^{-1}\mathrm{s}^{-2}\quad .$$

### 8.10.2 1. Keplersches Gesetz

Die Planeten bewegen sich auf Ellipsen, in deren einem Brennpunkt die Sonne steht.

```python
%matplotlib inline
%config InlineBackend.close_figures=False 

import ipywidgets as widgets
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['font.size'] = 15

t = np.linspace(0., 2.*np.pi, 1000)
plt.figure(1, figsize=(8, 8));

def update_plot(a,b):
    e=np.sqrt(1-b**2/a**2)
    c=a*e
    ax=plt.gca()
    ax.clear()
    x=a*np.cos(t)
    y=b*np.sin(t)
    plt.plot(x,y)
    ## Foki
    plt.plot(c,0,'go',markersize=15,alpha=0.5)
    plt.plot(-c,0,'go',markersize=15,alpha=0.5)
    ax.arrow(0, 0, -c, 0, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.arrow(-c, 0, c, 0, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.text(-c/2,0.05,'f')
    
    ## semiaxis
    plt.plot([-a,a],[0,0],'k--',lw=1)
    plt.plot([0,0],[-b,b],'k--',lw=1)
    ax.arrow(0, 0, a, 0, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.arrow(a, 0, -a, 0, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.arrow(0, 0, 0, b, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.arrow(0, b, 0, -b, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.text(a/2,-0.1,'a')
    ax.text(-0.1,b/2,'b')
    
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
 
    plt.show();

w_a = widgets.FloatSlider(.8, min=.5, max =1., step=0.025, description=r'$a$:')
w_b = widgets.FloatSlider(.5, min=.2, max =0.5, step=0.025, description=r'$b$:')
widgets.interactive(update_plot, a=w_a, b=w_b)
```

### 8.10.3 3. Keplersches Gesetz

Die Quadrate der Umlaufzeiten der Planeten verhalten sich wie die dritten Potenzen ihrer großen Halbachsen.

$$\frac{T_1^2}{T_2^2}=\frac{a_1^3}{a_2^3}$$

oder 

$$\frac{T_1^2}{a_1^3}=\frac{T_2^2}{a_2^3}=\dots=\mathrm{konst.}$$

Wir beweisen das für Kreibahnen um die Sonne mit Massse $m_S$. Dazu betrachten wir zwei Planeten mit Masse $m_1, m_2$ und Radii $r_1, r_2$ und Umlaufzeit $T_1, T_2$.

$$\vec{F}_1=-\gamma\frac{m_S\,m_1}{r_1^2}\vec{e}_r$$

Auf der Kreisbahn ist die Radialkraft gleich der Anziehungskraft.

$$\vec{F}_{Rad,1}=-m_1\,\omega_1^2\,r_1\,\vec{e}_r$$


und damit

$$\gamma\frac{m_s}{r_1^2}=\omega_1^2\,r_1$$

Jetzt ersetzen wir noch $\omega_1=2\pi/T_1$ und erhalten

$$\frac{T_1^2}{r_1^3}=\underbrace{\quad 4\pi^2\frac{1}{\gamma \, m_S}\quad}_\mbox{Identisch für alle Planeten um die Sonne}$$


Es folgt also

$$\frac{T_1^2}{r_1^3}=\frac{T_2^2}{r_2^3}\quad .$$
