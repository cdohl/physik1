# 8. Statik und Dynamik des starren Körpers
## 8.1 Grundbegriffe

* starrer Körper
  * System von Punktmassen (PM)
  * zwischen den PM starre Abstände
  * Form bleibt erhalten
  
* Bewegung
  * Translation des Schwerpunktes
  * Rotation um Achsen

* Gesamtenergie der Bewegung?

  Benutzung von Freiheitsgraden
  * 3 Freiheitsgrade der Translation
  
    Kinetische Energie der Translation
  
    $$ E_{kin}=\frac{1}{2}m v^2=\frac{1}{2}m\left( v_x^2+v_y^2+v_z^2\right)$$
  
  * 3 Freiheitgrade der Rotation (um die möglichen Rotationsachsen)
    
    für  eine Rotation nur um eine Achse, z.B. $\parallel \vec{e}_x$ durch den Schwerpunkt $S$ 
    
    $$E_{rot}=\frac{1}{2}J_{s,x}\,\omega^2$$
    
    $J_{s,x}$ wird das Massenträgheitsmoment parallel zur $\vec{e}_x$-Richtung genannt.

## 8.2 Drehimpuls und Rotationsenergie

<img src="figures/8.2_1.png" width=280px>

Der Drehimpuls eines Körpers mit der Geschwindigkeit $\vec{v}$ und Masse $m$ (Impulse $\vec{p}$ ist gegeben durch

$$\vec{L}_A=\vec{r}\times(m\,\vec{v})=\vec{r}\times\vec{p}$$

Hier $\vec{r}$ ist der Ortsvektor zum Bezugspunkt $A$ ist. Der Drehimpuls $\vec{L}$ ist immer abhängig vom Bezugspunkt, und daher schreiben wir $\vec{L}_A$.

* Betrag von $\vec{L}$

  $$\left|\vec{L}\right|=\left|\vec{r}\right|\,\left|\vec{p}\right|\,\sin\left(\vec{r},\vec{p}\right)=\left|\vec{r}\right|\,\left|\vec{p}\right|\,\sin \gamma$$
  
  * $\left|\vec{L}\right|$ wird maximal wenn $\vec{r}\perp\vec{v}$
  
  * $\left|\vec{L}\right|$ wird minimal wenn $\gamma=0^\circ$ oder $\gamma=90^\circ$, also $\vec{r}\parallel\vec{v}$
  
  * Richtung von $\vec{L}$ mit der Rechten-Hand-Regel
  
  
* $\vec{L}_Z$ einer Punktmasse auf einer Kreisbahn


  <img src="figures/8.2_2.png" width=240px>
  
  Wir wählen als Bezugspunkt den Kreismittelpunkt $Z$ aus. Mit Hilfe der Rechten-Hand-Regel sehen Sie, dass der Drehimpuls aus der Bildebene zu Ihnen zeigt. Damit ist die Richtung identisch mit der Richtung der Drehachse $\vec{\omega}$. Der Betrag läßt sich leicht bestimmen, da $\vec{r}_Z\perp\vec{v}$ ist.

  \begin{eqnarray}
  \left|\vec{L}_Z\right|&=&m\,\left|\vec{r}_Z\times\vec{v}\right|\\
  &=&m\,r_Z\,v=m\,{r_Z}^2\,\omega
  \end{eqnarray}
  
  Da wir die Richtung von $\vec{L}_Z$ kennen können wir gleich
  
  $$\vec{L}_Z=m\,{r_Z}^2\,\vec{\omega}$$
  
  schreiben.
  
  $J_{A,\vec{\omega}}=m\,r^2$ ist das Massenträgheitsmoment bezüglch einer Achse $\vec{\omega}$ durch den Punkt $A$, dann ist der Drehimpuls
  
  $$\vec{L}=J_{A,\vec{\omega}}\,\vec{\omega}$$
  
  
  Jetzt beschreiben wird die Rotationsenergie des einen Massenpunktes $m$ durch die kinetische Energie des Systems:
  
  $$E_{kin}=\frac{1}{2}m\,v^2=\frac{1}{2}m\,r_Z^2\,\omega^2$$
  
  Hier ist die Bewegung eine reine Rotation, d.h. die kinetische Energie besteht nur aus der Rotationsenergie
  
  $$E_{rot}=E_{kin}$$
  
  und die Konstante in der Rotationsenergie ist $J_{Z,\vec{\omega}}=m\,r_Z^2$.
  
  $$E_{rot}=\frac{1}{2}J_{Z,\vec{\omega}}\,\omega^2$$
  
  $J_{Z,\vec{\omega}}$ ist das Massenträgheitsmoment der Rotationsbewegung um eine Achse die durch den Punkt $Z$ in Richtung $\vec{\omega}$ geht.

## 8.3 Massenträgheitsmoment 

Bestimmung des Massenträgheitsmomentes für einen ausgedehnten Körper

<img src="figures\8.2_3.png" width=300px>

Dazu brauchen wir eine Drehachse $\vec{\omega}$ und einen Drehpunkt $A$ durch den die Drehachse geht. Dann zerlegen wir den Körper in Massenelemente $\Delta m_i$

$$ m=\sum_i^N \Delta m_i$$

Dann ist das Massenträgheitsmoment

$$J_{A,\vec{\omega}}=\sum_i^N\,\Delta m_i\,{r_{A_i}}^2\quad .$$

Für einen kontinuierlichen Körper lassen wir die Massenelemente $\Delta m_i\rightarrow 0$ gehen und benutzen infintisimale kleine Massen).

$$\rightarrow\sum\rightarrow\int$$

$$J_{A,\vec{\omega}}=\int {r_A}^2 \, \mathrm{d}m$$


#### Beispiel für einen Zylinder

mit konstanter Dichte, der um die Symmetrieachse $\vec{\omega}=\omega\,\vec{e}_z$ durch das Zentrum $S: x=y=z=0$ rotiert:

<img src="figures\8.3_2.png" width=200px>

$$J_{S,z}=\int r^2 \,\mathrm{d}m\quad .$$

Bei konstanter Dichte ist $\mathrm{d}m=\rho\,\mathrm{d}V$ und

$$J_{S,z}=\rho\int r^2 \,\mathrm{d}V\quad .$$

Für die Integration über $\mathrm{d}V$ muß ein geeignetes Volumenelement genommen werden. Hier bieten sich zylindrische Koordinaten an. Diese kann man leicht verstehen:

<img src="figures\8.3_3.png" width=250px>

Das Flächenelement $\mathrm{d}A$ eines Querschnitts des Zylinders ist

$$\mathrm{d}A=\underbrace{r\,\mathrm{d}\varphi}_{\mathrm{Bogenlänge}}\,\mathrm{d}r$$

Das Volumen beruücksichtigt dann nur noch die Tiefe in $z$-Richtung:

$$\mathrm{d}V=r\,\mathrm{d}\varphi\,\mathrm{d}r\,\mathrm{d}z$$

$$J_{S,z}=\rho\int r^2 \,\mathrm{d}V\quad=
\rho\int_0^h\int_0^R\int_0^{2\pi}r^2\,r\,\mathrm{d}\varphi\,\mathrm{d}r\,\mathrm{d}z$$

Integration über $\mathrm{d}\varphi$ ergibt $2\pi$

$$J_{S,z}=2\pi\,\rho\int_0^h\int_0^R r^3\mathrm{d}r\,\mathrm{d}z$$

Integration über $\mathrm{d}z$ ergibt $h$

$$J_{S,z}=2\pi\,\rho h \int_0^R r^3\mathrm{d}r$$

Und zum Schluss über $r$

$$J_{S,z}=2\pi\,\rho h \left[\frac{r^4}{4}\right]_0^R=\underbrace{\rho \pi R^2 h}_{\rho\,V=m} \,\frac{1}{2}R^2=\frac{1}{2}m R^2$$



#### Rollbewegung
Eine Rollbewegung besteht aus der Translation des Massenzentrums und einer Rotation um eine Achse durch den Schwerpunkt des Rollkörpers. Beim idealen Rollen ohne zu gleiten sind beide Geschwindigkeiten durch die Beziehung $v=\omega\,R$ gegeben.
<img src="figures\8.3_4.png" width=350px>


#### Geschwindigkeit eines rollenden Zylinders
Wir können die Geschwindigkeit eines Zylinders, der eine schiefe Ebene hinab rollt leicht mit dem Energieerhaltungssatz bestimmen. Dazu nehmen wir an, dass keine Reibungsarbeit geleistet wird (der Kontaktpunkt zwischen Zylinder und Ebene ruht).

<img src="figures\8.3_5.png" width=350px>


Zu Beginnn sei der Zylinder in Ruhe, dann ist 

$$E_{pot}=m g h$$

Danach wird er schneller. Wir haben zwei kinetische Energien durch die Translation und die Rotation:

$$E_{kin}=\frac{1}{2} m v^2+\frac{1}{2}J_S\,\omega^2$$

Nachdem der Zylinder die Höhe $h$ heruntergerollt ist, haben wir die potentielle Energie in kinetische Energie umgewandelt, also ist die Geschwindigkeit

$$v=\sqrt{\frac{2 g h}{1+\frac{J_s}{m R^2}}}$$

Im Falle eines Zylinders mit $J_S=\frac{1}{2}mR^2$

erhalten wir eine Geschwindigkeit von 

$$v=\sqrt{\frac{4}{3}g\,h}<\sqrt{2 g h}$$

Die Geschwindigkeit ist kleiner als die Geschwindigkeit eines Massenpunktes, der eine Höhe $h$ frei fällt, nämlich $v=\sqrt{2 g h}$. Warum?


## 8.4 Satz von Steiner 

<img src="figures\8.4_1.jpg" width=250px>

Das Trägheitsmoment $J_A$ eines Körpers um die Achse $A$ setzt sich additiv aus dem Trägheitsmoment $J_S$ bezüglich des Schwerpunktes und dem Trägheitsmoment $m\,a^2$. Dabei bezeichnet $a$ den Abstand der beiden parallelen Achsen und $m$ die gesamte Masse des Körpers. <b>Wichtig</b> Beide Achsen müssen parallel sein.





## 8.5 Dynamisches Grundgesetz der Rotation

* Bei der Translation führt eine Kraft zu einer Impulsänderung

  $$ \vec{F}=\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}$$

* Wann ändert sich der Drehimpuls?
  Dazu bestimmmen wir die zeitliche Änderung von $\vec{L}$:
  
  \begin{eqnarray}
  \frac{\mathrm{d}\vec{L}}{\mathrm{d}t}&=&\frac{\mathrm{d}\,(\vec{r}\times\vec{p})}{\mathrm{d}t}\\
  &=&\underbrace{\frac{\mathrm{d}\vec{r}}{\mathrm{d}t}\times\vec{p}}_{\vec{v}\times\vec{p}=0}+\vec{r}\times\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}\\
  \frac{\mathrm{d}\vec{L}}{\mathrm{d}t}&=&\vec{r}\times\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}=\vec{r}\times\vec{F}
  \end{eqnarray}

* Die zeitliche Änderung des Drehimpulses wird Drehmoment $\vec{M}$ genannt

  $$\vec{M}=\vec{r}\times\vec{F}$$
  
  Der Drehimpuls ist auf einen Körper ist
  
  $$\vec{L}=J_A\,\vec{\omega}$$
  
  und damit ist 
  
  $$\vec{M}=\frac{\mathrm{d}\vec{L}}{\mathrm{d}t}=
  \frac{\mathrm{d}J_A}{\mathrm{d}t}\vec{\omega}+J_A\,\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}$$

  Die Winkelbeschleunigung ist 
  
  $$\vec{\alpha}=\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}$$
  
  und für den Fall eines konstanten Trägheitsmoments erhalten wir das *Grundgesetz der Rotation*
  
  $$\vec{M}=J_A\,\vec{\alpha}$$


## 8.6 Drehimpulserhaltungssatz

* abgeschlossenes physikalisches System

  $$\vec{M}=\sum_{i=1}^N\vec{M}_i$$
  
  Es wirkt keine äußere Kraft an, deshalb ist $\vec{M}=0$ und da $\vec{M}=\frac{\mathrm{d}\vec{L}}{\mathrm{d}t}$ ist $\vec{L}=\mathrm{konst}$.
  
* für mehrere Körper im System

  $$\vec{L}=\sum_{i=1}^N\vec{L}_i$$
  
  
* Beispiel 1: $\vec{L}=0$

  für zwei Körper:
  
  $$\vec{L}=J_1\vec{\omega}_1+J_2\vec{\omega}_2=0$$
  
  $$\vec{\omega}_2=-\frac{J_1}{J_2} \vec{\omega}_1$$
  
  
* Beispiel 2: $\vec{L}=\mathrm{konst.}\ne0$

  Ausgangszustand
  $$\vec{L}= J_1\vec{\omega}_1$$
  
  Endzustand
  $$\vec{L}= J_2\vec{\omega}_2$$
  
  $$\vec{\omega_2}=\frac{J_1}{J_2}\vec{\omega}_1$$
