### 3.5 Kreisbewegung

#### Ebene Polarkoordinaten

Radius $r=\left|\vec{r}\right|$

Drehwinkel $\varphi$


#### Kreisbahn

$\left|\vec{r}\right|=r=\mathrm{konst.}$

Drehwinkel: $\varphi=\varphi(t)$

$$x(t)=r\,\cos \left(\varphi(t)\right)$$

$$y(t)=r\,\sin \left(\varphi(t)\right)$$

* Winkelgeschwindigkeit

  $$\omega=\frac{\mathrm{d}\varphi}{\mathrm{d}t}$$
  
  $$\left[\varphi\right]=1\,\mathrm{rad}$$
  
  $$\left[\omega\right]=\frac{\mathrm{rad}}{\mathrm{s}}\stackrel{\frown}{=}\frac{1}{\mathrm{s}}$$
  
* Winkelbeschleunigung

  $$\alpha=\frac{\mathrm{d}\omega}{\mathrm{d}t}=\dot{\omega}=\frac{\mathrm{d}^2\varphi}{\mathrm{d} t^2}=\ddot{\varphi}$$
  
#### Kreisbewegung

* gleichförmige Kreisbewegung: $\omega=\mathrm{konst}$ und $\alpha=0$
* ungleichförmige Kreisbewegung:  $\omega\ne\mathrm{konst}$ und $\alpha\ne0$


* Beispiel: gleichförmige Kreisbewegung
  \begin{eqnarray}
  \alpha = \frac{\mathrm{d}\omega}{\mathrm{d}t} &=& 0\\
  \int \mathrm{d}\omega &=& \int 0 \, \mathrm{d}t\\
  \omega &=& C_1=\mathrm{konst.}
  \end{eqnarray}
  
  Für die Bestimmung des zeitabhängigen Drehwinkel müsssen wir noch einmal integrieren:
  
  \begin{eqnarray}
  \omega&=&\frac{\mathrm{d}\varphi}{\mathrm{d}t}\\
  \mathrm{d}\varphi&=&\omega\,\mathrm{d}t\\
  \int\mathrm{d}\varphi&=&\int \omega\,\mathrm{d}t\\
  \varphi(t)&=&\omega\,t + \varphi_0 
  \end{eqnarray}
  
  $\varphi_0$ ist der Nullphasenwinkel, aka $\varphi_0=\varphi(t=0)$.
  
  ⇒ Periodendauer für einen vollen Umlauf $T$
  
  $$\omega=\frac{2\,\pi}{T}$$
  
  ⇒ Frequenz (Anzahl der Umläufe pro Zeit)
  
  $$f=\frac{1}{T}=\frac{\omega}{2\,\pi}$$
  
  Zusammenhang zwischen Winkelgeschwindigkeit und Frequenz
  
  $$\omega=2\,\pi\,f$$

```python
%matplotlib inline
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import matplotlib.patches as patches
mpl.rcParams['font.size'] = 15

R=1
alpha=np.linspace(0,2*np.pi,200)



def update_plot(phi):
    
    phirad=phi*np.pi/180.
    plt.figure(1, figsize=(6, 6))
    ax=plt.gca()
    plt.plot(R*np.cos(alpha),R*np.sin(alpha))
    ax.axhline(color='k',linewidth=1);ax.axvline(color='k',linewidth=1)
    ax.text(R*1.3,-0.1,r'$x$');ax.text(-0.1,R*1.3,r'$y$')
    ax.set_xticks([]);ax.set_yticks([])
    ax.spines['top'].set_visible(False);ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False);ax.spines['left'].set_visible(False)
    
    #velocity vector
    ax.arrow(R*np.cos(phirad), R*np.sin(phirad), -.7*np.sin(phirad), .7*np.cos(phirad), lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.text(R*np.cos(phirad)-.35*np.sin(phirad),R*np.sin(phirad)+.35*np.cos(phirad),r'$\vec{v}$',fontsize=18)
    #angle
    arc = patches.Arc((0.,0), R, R, 1.,0.,phi,linewidth=2.)
    ax.add_patch(arc)
    ax.text(R*0.3,0.1,r'$\varphi$')
    plt.plot([0,R*np.cos(phirad)],[0,R*np.sin(phirad)],'k')
    plt.plot([R*np.cos(phirad),R*np.cos(phirad)],[0,R*np.sin(phirad)],'k--')
    plt.plot([0,R*np.cos(phirad)],[R*np.sin(phirad),R*np.sin(phirad)],'k--')
    ax.text(.5*R*np.cos(phirad)-.05*np.sin(phirad),.5*R*np.sin(phirad)+.05*np.cos(phirad),r'$\vec{r}$')
    #axis limits
    ax.set_xlim([-1.3*R, 1.3*R])
    ax.set_ylim([-1.3*R, 1.3*R])
    plt.show();
    
    
tv = widgets.FloatSlider(value=30., min=0., max =360., step=15.,description=r'$\phi$:');
widgets.interactive(update_plot, phi=tv)
```

#### Kreisbewegung ist beschleunigte Bewegung

Jede Kreisbewegung ist eine beschleunigte Bewegung, da sich mindestens die Richtung von $\vec{v}$ ständig ändert.  
<img src="figures/3.5_1.png" width="300">

Das sieht man deutlich wenn man die beiden Vektoren $\vec{v}_1$ und $\vec{v}_2$ vergleicht

<img src="figures/3.5_2.png" width="100">


$$\vec{v}_2=\vec{v}_1+\Delta \vec{v}$$

Es müssen zwei Beschleunigungen (Änderungen des Geschwindigkeitsvektors) berücksichtigt werden. Diese lassen sich sehr elegant in Zylinderkoordinaten beschreiben:

#### Einführung von Zylinderkoordinaten

Der Einheitsvektor $\vec{e}_r$ läuft radial, also ist

$$\vec{r}=r\,\vec{e}_r$$

und der Einheitsvektor $\vec{e}_t$ ist tangential zum Kreis

$$\vec{v}=v\,\vec{e}_t\quad .$$

<img src="figures/3.5_3.png" width="200">

Der dritte Einheitsvektor ist senkrecht zur Ebene der Drehbewegung: $\vec{e}_z$

$$\vec{e}_z\perp\vec{e}_r\perp\vec{e}_t$$

$$\vec{e}_z\times\vec{e}_r=\vec{e}_t$$


* für eine gleichförmige Kreisbewegung, also $\omega=\mathrm{konst.}$

  $$\left|\vec{v}\right|=\frac{2\pi\,r}{T}$$

  Merke Bahnlänge pro Zeit!

* für eine ungleichförmige Kreisbewegung

  $$\left|\vec{v}\right|=v\ne\mathrm{konst.}$$
  
  und/oder
  
  $$\omega\ne\mathrm{konst.}$$
  
  Die Gesamtbeschleunigung ergibt sich aus der Tangentialbeschleunigung (Änderung von $\omega$) und der Radialbeschleunigung (Änderung der Richtung von $\vec{v}$).
  
  $$\vec{a}_t=\frac{\mathrm{d}\omega}{\mathrm{d}t}r\,\vec{e}_t=\dot{\omega}\, r\,\vec{e}_t=\alpha\, r\,\vec{e}_t$$
  
  $$\vec{a}_r=-\frac{v^2}{r}\,\vec{e}_r$$
  
In Kürze:

  * Tangentialbeschleunigung $\vec{a}_t$:
  ändert die Winkelgeschwindigkeit/Rotationsgeschwindigkeit
  
  * Radialbeschleunigung $\vec{a}_r$:
  ändert die Richtung von $\vec{v}$
  
  $$\vec{a}=\vec{a}_t+\vec{a}_r$$
  
  <img src="figures/3.5_4.png" width="175">
  
  Wenn $\vec{a}_t=0$, dann ist $\dot{\omega}=0$ und $\vec{a}=\vec{a}_r$
