## 4.3 Kräfte

### 4.3.1 Elastische Kräfte, Federkraft

Die lineare Feder erzeugt eine Rückstellkraft, die linear zur Auslenkung $x$ ist. Die Richtung ist entgegen der Auslenkung was das negative Vorzeichen in der Gleichung (4.3.1) erklärt.

<img src="figures/4.3_1.png" width=270px>

$$\vec{F}_F=-k\,x\,\vec{e}_x\tag{4.3.1}$$

Eine elastische Kraft ist wegunabhängig (nur von der Auslenkung). Diese Art von Kräfte nennt man konservative Kräfte.

### 4.3.2 Reibungskräfte

* Haftreibung ($\vec{v}=0$), d.h. das Objekt ist in Ruhe)

<img src="figures/4.3_2.png" width=270px>

Die Haftreibung widersetzt die angreifende Kraft bis zu einem maximalen Wert. Solange dieser maximaler Wert nicht erreicht ist, bleibt der Körper in Ruhe.

$$\vec{F}_{R}+\vec{F}=0$$

Auf den Körper wirkt eine Normalkraft $\vec{F}_N=m\,g\,\vec{e}_y$, und nur diese bestimmt die maximale Haftreibung über einen Haftreibungskoeffizienten $\mu_0$.

$$F_{R,max}=\mu_0\,F_N$$

Die maximale Haftreibung und damit den Reibungskoeffizienten kann man bequem auf einer schiefen Ebene bestimmen

<img src="figures/4.3_3.png" width=270px>

Die Kraft die den Körper nach unten bewegt ist 
$$F=m\,g\sin\alpha$$

und die Normalkraft

$$F_N=m\,g\,\cos\alpha\quad .$$

Die Haftreibungskraft ist maximal $F_{R,max}=-\mu_0\,F_N$. Das heisst ist die Kraft an dem Körper grösser als $\left|F_{R,max}\right|$, dann rutsch der Körper. Das Rutschen startet also wenn

$$\left|F\right|=\left|F_{R,max}\right|$$

$$m\,g\,\sin\alpha=\mu_0\,F_N=\mu_0\,m\,g\cos\alpha$$

$$\tan\alpha=\mu_0$$

* Gleitreibungskraft

Wenn der Körper mit einer Geschwindigkeit $\vec{v}=v\,\vec{e}_v$ bewegt, wirkt auf den Körper die Gleitreibungskraft. Diese Kraft wirkt entgegen der Richtung der Geschwindigkeit $\vec{e}_v$.  

<img src="figures/4.3_4.png" width=270px>

$$\vec{F}_{GL}=-\mu_{GL}\left|\vec{F}_N\right|\vec{e}_v$$

den Einheitsvektor kann man auch so ausdrücken

$$\vec{e}_v=\frac{\vec{v}}{\left|\vec{v}\right|}$$

und damit

$$\vec{F}_{GL}=-\mu_{GL}\left|\vec{F}_N\right|\frac{\vec{v}}{\left|\vec{v}\right|}\quad .$$

$F_{GL}$ ist unabhängig von dem Betrag von $\vec{v}$.
