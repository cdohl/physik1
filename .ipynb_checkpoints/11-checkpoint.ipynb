# 11. Kinetische Gastheorie

* makroskopische Eingenschaften eines Gases aus mikroskopischen Modellen herleiten

* Phänomenologische Thermodynamik
  * Makrozustand eines Systems soll durch Zustandsgrößen beschrieben werden
    * Druck $p$
    * Volumen $V$
    * absolute Temperatur $T$
    
## 11.1 Grundbegriffe/Ideales Gas

* n - Stoffmenge $[n]=1\,$mol

  für $n=1\,$mol ⇒ $N=6\cdot 10^{23}\,$Teilchen
  $N = n\, N_A$ mit der Avogadro-Konstante
  
  $N_A=6.023\cdot 10^{23}\,$mol$^{-1}$
  
* Die Masse aller Teilchen $m$
  
  $$m=n\,M$$
  
  $M$: molare Masse, $[M]=1\frac{\mathrm{g}}{\mathrm{mol}}$
  
* Die Masse eines Gasteilchens $\mu$

  $$\mu=\frac{M}{N_A}=\frac{m}{N}$$
  
* Boltzmann-Konstante

  $$ k =1.38\,\cdot 10^{-23}\,\frac{\mathrm{J}}{\mathrm{K}}$$
  
* Universelle Gaskonstante

  $${\cal R}=k\,N_A$$
  
  $${\cal R}=8.31 \,\frac{\mathrm{J}}{\mathrm{mol}\cdot\mathrm{K}}$$
  
*Modellvorstellung: Ideales Gas*

* Atome und Gasmoleküle sind Punktmassen (kein Eigenvolumen)
* Stöße zwischen den Gasatomen und den Gefäßwänden sind vollkommen elastisch
* Es gibt keine weiteren Wechselwirkung
* Realisierbar durch
  * große Gefäßvolumina mit kleinen Teilchenanzahlen
  * durch hohe Temperaturen


## 11.2 Maxwell-Boltzmannsche Geschwindigkeitsverteilung

* Alle Teilchen haben unterschiedliche Geschwindigkeiten
* Nur angebbar die Wahrscheinlichkeit ein Teilchen in einem Geschwindigkeitsintervall zwischen $v$ und $v+\mathrm{d}v$ zu finden.
* Die Beschreibung der Verteilungen der Geschwindigkeiten (äußere Bedingungen) erfolgt durch eine Verteilungsfunktion

  $$f(v)$$
  
* Betrachtung von $N$ Teilchen ⇒ dann befinden sich $\mathrm{d}N$ im Intervall $\mathrm{d}v$

  $$\mathrm{d}N=N\,f(v)\,\mathrm{d}v\tag{11.1}$$
  
<img src="figures\11.1_1.png" width=350px>
  
  $$\mathrm{d}N=N\,\underbrace{4\pi v^2\left(\frac{\mu}{2 k T}\right)^{3/2}\exp\left(-\frac{\mu v^2}{2 k T}\right)}_{f(v)}\,\mathrm{d}v$$
  
  
  
* Thermische Energie: $E_T=k\,T$

* Kinetische Energie: $E_{kin}=\frac{1}{2}\mu v^2$

$$\exp\left(-\frac{\mu v^2}{2 k T}\right)=\exp\left(\frac{E_{kin}}{E_T}\right)$$

und damit wird die Verteilungsfunktion

$$f(v)=4\pi v^2\left(\frac{\mu}{k t}\right)^{3/2}\exp\left(\frac{E_{kin}}{E_T}\right)$$

für kleine $v$ überwiegt $v^2$, d.h. $f(v)\propto v^2$

und für große $v$ ist der exponentielle Term dominant $f(v)\propto \exp\left(-\frac{\mu v^2}{2 k T}\right)$.

```python
%matplotlib inline
%config InlineBackend.close_figures=False 

import ipywidgets as widgets
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['font.size'] = 15

v = np.linspace(0., 4., 1000)
vw=2**(1/2.)
vsqr=3**(1/2.)
vbar=np.sqrt(8/np.pi)
f=np.sqrt(2/np.pi)*v**2*np.exp(-(v**2)/2.)
fv2=np.sqrt(2/np.pi)*v**2
fexp=np.exp(-((v-vw)**2)/2.)/5.

plt.figure(1, figsize=(8, 6));
ax=plt.gca()
ax.clear()
ax.plot(v,f,'k',label="Maxwell Boltzmann")
ax.plot(v,fv2,'--',label=r'$\propto v^2$')
ax.plot(v[600:],fexp[600:],'--',label=r'$\propto \exp(-\frac{\mu v^2}{2kT})$')
ax.plot([vw,vw],[0,1.],'k-.')
ax.plot([vbar,vbar],[0,1.],'k-.')
ax.plot([vsqr,vsqr],[0,1.],'k-.')
ax.legend()
ax.set_ylabel(r'$f(v)$')
ax.set_xlabel('Geschwindigkeit $v$')
#ax.set_xlim(0, 1.)
ax.set_ylim(-0.05, 1.)
plt.xticks([vw,vbar,vsqr],[r'$v_W$',r'$\overline{v}$',r'$\sqrt{\overline{v^2}}$'])
plt.yticks([],[])
plt.show();
```

* Eigenschaften von $f(v)$
  * $f(v)\ge 0$
  * $f(v=0)=0$
  * für kleine $v$ ⇒ $f(v)\propto v^2$
  * für große $v$ ⇒ $f(v)\exp(-v^2)$
  * $v\rightarrow\infty$ ⇒ $f(v)\rightarrow 0$
  * Normierbarkeit (Flächeninhalt under der Kurve)
    $$\int_0^\infty f(v)\,\mathrm{d}v=1$$
  * physikalische Einflußgrößen $\mu$ und $T$
  * Charakteristische Geschwindigkeiten $v_W$,$\bar{v}$,$\sqrt{\bar{v}}$
  
    * $v_W$ wahrscheinlichste Geschwindigkeit
    
      $$ \left.\frac{\mathrm{d} f(v)}{\mathrm{d}v}\right|_{v=v_w}=0$$
      $$v_W=\sqrt{2}\,\sqrt{\frac{k T}{\mu}}$$
    
    * $\overline{v}$ mittlere Geschwindigkeit
    
      $$\overline{v}=\int_0^\infty v \, f(v) \mathrm{d} v$$
      $$\overline{v}=\sqrt{\frac{8}{\pi}}\,\sqrt{\frac{k T}{\mu}}$$
      
    * $\sqrt{\overline{v^2}}$ mittlere quadratische Geschwindigkeit
    
      $$\sqrt{\overline{v^2}}=\sqrt{\int_0^\infty v^2 \, f(v) \mathrm{d} v}$$
      $$\sqrt{\overline{v^2}}=\sqrt{3}\,\sqrt{\frac{k T}{\mu}}$$
      
  * Einfluss der Temperatur
    $v_W\propto\sqrt{T}$, d.h. wenn sich die Temperatur vervierfacht verdoppelt sich die wahrscheinlichste Geschwindigkeit.
    

```python
%matplotlib inline
%config InlineBackend.close_figures=False 

import ipywidgets as widgets
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['font.size'] = 15

v = np.linspace(0., 4., 1000)

def update_plot(T):
    vw=2**(1/2.)*np.sqrt(T)
    vsqr=3**(1/2.)*np.sqrt(T)
    vbar=np.sqrt(8/np.pi)*np.sqrt(T)
    f=np.sqrt(2/np.pi)*v**2*np.exp(-(v**2)/(2.*T))*T**(-3/2.)
    fw=np.sqrt(2/np.pi)*vw**2*np.exp(-(vw**2)/(2.*T))*T**(-3/2.)
    fsqr=np.sqrt(2/np.pi)*vsqr**2*np.exp(-(vsqr**2)/(2.*T))*T**(-3/2.)
    fbar=np.sqrt(2/np.pi)*vbar**2*np.exp(-(vbar**2)/(2.*T))*T**(-3/2.)
    plt.figure(1, figsize=(8, 6));
    ax=plt.gca()
    ax.clear()
    ax.plot(v,f,'k',label="Maxwell Boltzmann")
    ax.plot([vw,vw],[0,fw],'k-.')
    ax.plot([vbar,vbar],[0,fbar],'k-.')
    ax.plot([vsqr,vsqr],[0,fsqr],'k-.')
    ax.legend()
    ax.set_ylabel(r'$f(v)$')
    ax.set_xlabel('Geschwindigkeit $v$')
    #ax.set_xlim(0, 1.)
    ax.set_ylim(-0.05, 1.)
    plt.xticks([vw,vbar,vsqr],[r'$v_W$',r'$\bar{v}$',r'$\sqrt{\bar{v^2}}$'])
    plt.yticks([],[])
    plt.show();
    
Tw = widgets.FloatSlider(value=.5, min=.01, max =2., step=0.1, 
                              description=r'$T$:')
widgets.interactive(update_plot, T=Tw)    
```

## 11.3 Gleichverteilungssatz und Innere Energie

* eine Gassorte ($\mu=\mathrm{konst.}$)
* die kinetische Energie der Translation für das i-te Teilchen in der klassischen Mechanik 
  $$E_{kin,i}=\frac{1}{2}\mu\,{v_i}^2$$

* Thermodynamik ⇒ mittlere kinetische Energie von Teilchen

  $$\overline{E_{kin}}=\frac{1}{N}\int_0^N\frac{1}{2}\mu\,v^2\mathrm{d}N$$
  $$ =\frac{1}{2}\mu\,\,\underbrace{\frac{1}{N}\int_0^N v^2\mathrm{d}N}_{\overline{v^2}}$$
  
  Das wird aus Gleichung (11.1) mit $\mathrm{d}N=N f(v) \mathrm{d}v$ ersichtlich. Damit ist dann
  
  $$\overline{E_{kin}}=\frac{1}{2}\mu\overline{v^2}=\frac{3}{2}\mu\frac{k\,T}{\mu}=\frac{3}{2} k\,T$$
  
  für die drei äquivalente Raumrichtungen mit $v_x^2$, $v_y^2$ und $v_z^2$
  
  $$\overline{E_{kin}}=\frac{1}{2} \mu\,\underbrace{\overline{v_x^2+v_y^2+v_z^2}}_\overline{v^2}$$
  
  Auf jede Raumrichtung entfällt im Mittel eine kinetische Energie von $\frac{1}{2}kT$. 
  
  Bewegungsrichtung entspricht einem Freiheitsgrad der Bewegung und gibt die Möglichkeit Energie zu speichern.
  
* Gleichgewichtssatz
  Auf jeden Freiheitsgrad entfällt im Mittel die Energie von $\frac{k\,T}{2}$

* Freiheitsgrade
  * Translation
  * Rotation
  * Schwingungen (bei hohen Temperaturen)
  
* jedes Gasteilchen besitzt drei Freiheitsgrade der Translation
  $$f_T=3$$
* Rotationsfreiheitsgrade
  * 1-atomige (Edelgase) $f_R=0$ da Punktmasse
  * 2-atomige $f_R=2$
  <img src="figures/11.3_1.png" width=180px>
  * mehr als zwei Atome $f_R=3$
  <img src="figures/11.3_2.png" width=180px>
  
* Ein Molekül trägt die Gesamtenergie

  $$\overline{E_{ges}}=\left(f_T+f_R+\dots\right) \frac{k T}{2}=f_{ges} \frac{k T}{2}$$
  
* für ein Gas mit $N$ Teilchen ⇒ Zusammenfassung von $E_{ges}$ zur Inneren Energie $U$

  $$U=N\,\overline{E_{ges}}=n N_A \frac{k T}{2} f_{ges}$$
  und da ${\cal R}=N_A\,k$ ist
  
  $$U=n\,f_{ges}\frac{{\cal R}\,T}{2}$$

## 11.4 Grundgleichung der kinetischen Gastheorie

* Der Druck als Zustandsgröße

<img src="figures/11.4_1.png" width=260px>

* Druck: $p=\frac{F_x}{A}$ ist die Wechselwirkungskraft bei Reflexion

* Impulsänderung an der Gefäßwand ($A$)

  $$F_x=\frac{\Delta p_x}{\Delta t}$$
  
  $$\Delta p_x=2\mu\,v_x$$
  
* ein Teilchen stößt alle $\Delta t=\frac{2L}{v_x}$ mit der Gefäßwand $A$
  
  $$F_X=\frac{\mu\,v_x^2}{L}$$
  
* Einsetzen in Druck

  $$p=\frac{\mu \, v_x^2}{A L}$$
  
  $$p=\frac{\mu \, v_x^2}{V}$$

* Mittelung über alle Teilchen

  $$\overline{p}=\frac{\mu \, \overline{v_x^2}}{V}$$
  
  und da
  $$\overline{v_x^2}=\overline{v_y^2}=\overline{v_z^2}=\frac{1}{3}\overline{v^2}$$
  ⇒ 
  $$\overline{p}=\frac{\mu\,\overline{v^2}}{3 V}$$
  
* für alle $N$ Teilchen ergibt sich ein Gesamtdruck

  $$p_{ges}=N\,\overline{p}=\frac{N \,\frac{1}{3}\mu\,\overline{v^2}}{V}$$
  
* im folgenden $p_{ges}=p$ und mit $\overline{v^2}=3\frac{k T}{\mu}$

  $$p=\frac{N}{V}\,k\,T=\frac{n \, N_A\, k\,T}{V}$$ 
  
  und mit Hilfe der idealen Gaskonstante ${\cal R}=N_A k$ erhalten wir das ideale Gasgesetz
  
  $$p\,V=n\,{\cal R}\,T$$
