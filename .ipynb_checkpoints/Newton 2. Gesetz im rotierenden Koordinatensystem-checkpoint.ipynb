
## Herleitung der Newtonschen Gesetze im Rotierenden Koordinatensystem

<img src="figures/add1_KinematikKreisbewegung.png" width=150px>

Der Geschwindigkeitsvektor bei der kreisförmigen Bewegung mit Winkelgeschwindigkeit $\omega$ und Drehachsenorientierung $\frac{\vec{\omega}}{\omega}$ ist gegeben durch

$$\vec{v}=\vec{\omega}\times\vec{r}\quad.$$

Durch die Drehung ändert sich der Vektor $\vec{r}$ kontinuierlich. Ein mitdrehendes kartesiches Koordinatensystem (hier gestrichen) mit Einheitsvektoren ${\vec{e}_x}',{\vec{e}_y}',{\vec{e}_z}'$ order kurz ${\vec{e}_i}'$ mit $i=1,2,3$ ändern ebenfalls seine Richtung. Deren Geschwindigkeit $\frac{\mathrm{d}\vec{e}'_i}{\mathrm{d}t}=\dot{{\vec{e}_i}'}$ ist somit

$$\dot{{\vec{e}_i}'}=\vec{\omega}\times{\vec{e}_i}'\quad.\tag{1}$$

Ganz allgemein kann damit eine Geschwindigkeit eines Vektors im inertialen (nichtrotierendem) System $\vec{f}(t)=f_x\vec{e}_x+f_y\vec{e}_y+f_z\vec{e}_z$
 schreiben
 
$$\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t} = 
\frac{\mathrm{d}\,f_x}{\mathrm{d}t}\vec{e}_x+
\frac{\mathrm{d}\,f_y}{\mathrm{d}t}\vec{e}_y+
\frac{\mathrm{d}\,f_z}{\mathrm{d}t}\vec{e}_z$$

Beschreiben wir aber den Vektor $f(t)$ mit dem rotierendem Koordinatensystem müssen wir die Kettenregel anwenden, da die Einheitsvektoren auch sich zeitlich ändern.

$$\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t} = 
\frac{\mathrm{d}\,f_x}{\mathrm{d}t}{\vec{e}_x}'+
\frac{\mathrm{d}\,f_y}{\mathrm{d}t}{\vec{e}_y}'+
\frac{\mathrm{d}\,f_z}{\mathrm{d}t}{\vec{e}_z}'+
f_x\frac{\mathrm{d}{\vec{e}_x}'}{\mathrm{d}t}+
f_y\frac{\mathrm{d}{\vec{e}_y}'}{\mathrm{d}t}+
f_z\frac{\mathrm{d}{\vec{e}_z}'}{\mathrm{d}t}$$

Wir können mit obiger Gleichung (1) die rechte Seite umformen:

$$\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t} = 
\frac{\mathrm{d}\,f_x}{\mathrm{d}t}{\vec{e}_x}'+
\frac{\mathrm{d}\,f_y}{\mathrm{d}t}{\vec{e}_y}'+
\frac{\mathrm{d}\,f_z}{\mathrm{d}t}{\vec{e}_z}'+
\omega \times \left(\,
f_x\,{\vec{e}_x}'+ \,f_y\,{\vec{e}_y}'+ \,f_z\,{\vec{e}_z}'\right)$$

oder so
$$\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t} =
\left(\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t}\right)_r+\omega\times\vec{f}(t)$$
wobei $\left(\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t}\right)_r$ die Geschwindigkeit gemessesn im rotierenden Koordinatensystem ist.

Ganz allgemein sind die Geschwindigkeiten eines Vektors im Inertialsystem $\vec{v}_i$ mit dem rotierenden System $\vec{v}_r$ über die folgende mathematische Operation verbunden:

$$\vec{v}_i=\left(\frac{\mathrm{d}\vec{r}}{\mathrm{d}t}\right)_r+\omega\times\vec{r}=\vec{v}_r+\omega\times\vec{r}$$

Die Beschleunigung ist die zweite zeitliche Ableitung und kann formell als

$$\vec{a}_i=\left(\frac{\mathrm{d}^2\vec{r}}{\mathrm{d}t^2}\right)_i=
\left(\frac{\mathrm{d}\vec{v}}{\mathrm{d}t}\right)_i=
\left[\left(\frac{\mathrm{d}}{\mathrm{d}t}\right)_r+\omega\times\vec{r}\right]
\left[\left(\frac{\mathrm{d}\vec{r}}{\mathrm{d}t}\right)_r+\omega\times\vec{r}\right]
$$

geschrieben werden. Wenn wir die einzelnen Ableitung ausführen (gute Übung für die Kettenregel) erhalten Sie:

$$\vec{a}_i=\left(\frac{\mathrm{d}^2\vec{r}}{\mathrm{d}t^2}\right)_r+2\vec{\omega}\times\vec{v}_r+\omega\times\left(\omega\times\vec{r}\right)+\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}\times\vec{r}\tag{2}$$

Der erste Term auf der rechten Seite von Gl. (2) ist die gemessene Beschleunigung im rotierenden Koordinatensytem, also $\vec{a}_r$. Damit können wir diese Beschleunigung mit der Beschleunigung $\vec{a}_i$ im Inertialsystem vergleichen.

$$\vec{a}_r=\vec{a}_i-2\vec{\omega}\times\vec{v}_r-\omega\times\left(\omega\times\vec{r}\right)-\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}\times\vec{r}\tag{3}$$

Das bedeutet also, dass im rotierenden Koordinatensystem neben der Beschleunigung $\vec{a}_i$ die folgenden Beschleunigungen auftreten:

Zentrifugalbeschleunigung: $-\omega\times\left(\omega\times\vec{r}\right)$

Coriolisbeschleunigung: $-2\vec{\omega}\times\vec{v}_r=2\vec{v}_r\times\vec{\omega}$ 

und die Eulerbeschleunigung: $-\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}\times\vec{r}$

Diese drei zusätzlichen Beschleunigungen entsprechen den Trägheitskräften


Zentrifugalkraft: $\vec{F}_{Zf}=-m\omega\times\left(\omega\times\vec{r}\right)$

Corioliskraft: $\vec{F}_C=2\,m\,\vec{v}_r\times\vec{\omega}$ 

Eulerkraft: $\vec{F}_{E}=-m\,\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}\times\vec{r}$
