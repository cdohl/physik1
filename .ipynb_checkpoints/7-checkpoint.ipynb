## 7. Dynamik von Punktmassensystemen
### 7.1 Impulserhaltungssatz

Der Impuls eines Körpers ist 

$$\vec{p}=m\,\vec{v}$$

und die Einheit ist $\left[p\right]$=kg m s$^{-1}$.


Für eine Impulsänderung muss eine Kraft wirken:

$$\vec{F}=m\,\vec{a}=m\,\frac{\mathrm{d}\vec{v}}{\mathrm{d}t}=\frac{\mathrm{d}\left(m\,\vec{v}\right)}{\mathrm{d}t}=
\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}$$

D.h. wenn keine Kraft auf einen Körper wirkt bleibt der Impuls konstant. Wir können den Impulsbegriff auf ein Gesamtsystem erweitern. Der Gesamtimpuls ergibt sich aus der Verktoraddition der Einzelpulse $\vec{p}_i$, also

$$\vec{p}_{ges}=\sum_i \vec{p}_i=\sum_i m_i\vec{v}_i$$

Dort gilt:

<b> In einem abgeschlossenen System in dem nur innere Kräfte wirken, bleibt der Gesamtimpuls $p_{ges}$ ungeaändert.</b>

Inner Kräfte sind die Kräfte zwischen den Körpern des Gesamtsystems. Äussere Kräfte wirken von aussen auf das System. 


Beispiel:

Wir betrachten die Bewegung von einem geschlossenem System (reibungsfreie Auflage). Anfangs sind zwei Massen $m_1$ und $m_2$ mit einer Feder unter einer Federkraft verbunden über ein Seil. Die Massen sind in Ruhe, daher ist der Gesamtimpuls $p_{ges}=0$.

<img src="figures/7.1_1.png" width=450px>

Dann wird das Seil gekappt und die Massen bewegen sich auseinander. Es wirken keine äusseren Kräfte, der Gesamtimpuls bleibt erhalten.

$$p_{ges}=m_1\,\vec{v_1}+m_2\,\vec{v_2}=0\tag{7.1}$$

Die Bewegung ist eindimensional $\vec{v_1}=v_1\vec{e}_x$ und $\vec{v_2}=v_2\vec{e}_x$, daher gilt:

$$\left|\frac{v_1}{v_2}\right|=\frac{m_2}{m_1}$$

Wenn beide Massen gleich gross sind, bewegen sich $m_1$ und $m_2$ mit gleichem Betrag der Geschwindigkeit voneinander weg.

Wir können Gl. (7.1) weiter umformen:

$$m_1\,v_1+m_2\,v_2=p_{ges}\tag{7.2}$$

$p_{ges} war im obigen Beispiel Null, allgemein für ein System kann es aber jeden konstanten Wert annehmen.

Einführen der Definition von Geschwindigkeit in Gl. (7.2) ergibt:

\begin{eqnarray}
m_1\frac{\mathrm{d}x_1}{\mathrm{d}t}+m_2\frac{\mathrm{d}x_2}{\mathrm{d}t}&=&p_{ges}\\
\frac{\mathrm{d}}{\mathrm{d}t}\left(m_1\,x_1+m_2\,x_2\right)&=&p_{ges}\tag{7.3}
\end{eqnarray}

Jetzt teilen wir beiden Seiten von Gl. (7.3) mit der Summe der Masse $m_1+m_2$

\begin{eqnarray}
\frac{1}{m_1+m_2}\frac{\mathrm{d}}{\mathrm{d}t}\left(m_1\,x_1+m_2\,x_2\right)&=&\frac{p_{ges}}{m_1+m_2}\\
\frac{\mathrm{d}}{\mathrm{d}t}\underbrace{\left(\frac{m_1\,x_1+m_2\,x_2}{m_1+m_2}\right)}_{x_2}&=&\mathrm{konst.}\tag{7.4}
\end{eqnarray}

Der Ausdruck in der Klammer in Gl. (7.4) ist zeitlich konstant und wird daher eine *Erhaltungsgröße* genannt. $x_s$ ist eine Ortskoordinate (Einheit Länge) und gibt den  Massenschwerpunkt des abgeschlossenen Systems an. Ganz allgemein folgt aus der Impulserhaltung, dass <b>der Massenschwerpunkt $\vec{r}_s$ des abgeschlossenen Systems konstant ist.</b>

$$\vec{r}_s=\frac{\sum_i m_i\,\vec{r}_i}{\sum_i m_i}=\mathrm{konst.}$$
