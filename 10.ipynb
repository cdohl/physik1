## 10.2 Kontinuitätsgleichung

* Strömung durch eine Fläche $\vec{A}$: die Größe $\vec{v}\cdot\vec{A}$ gibt den Volumenstrom durch die Fläche $\vec{A}$ an. Die Fläche ist ein Vektor, der Betrag gibt die Größe der Fläche und der Vektor die Richtung an. 

<img src="figures/10.1_2.png" width=300px>

Der Volumenstrom $Q$ ist 

$$Q=\vec{v}\cdot\vec{A}=\underbrace{|\vec{v}|\,|\vec{A}|\,\cos(\vec{v},\vec{A})}_{A_\perp}=|\vec{v}|\,A_\perp= v\,A$$


* Kontinuitätsgleichung

<img src="figures/10.1_3.png" width=420px>

$$\Delta V_1=A_1\Delta x_1\quad\quad\quad\Delta V_2=A_2\Delta x_2$$

* Erhaltung des Massestromes durch die Fläche $A_1$ bzw. $A_2$

* Der Massestrom ist

$$\frac{\Delta m}{\Delta t}$$

Damit ist $$\Delta m_1=\rho_1\,\Delta V_1$$
und $$\Delta m_2=\rho_2\,\Delta V_2$$


$$\frac{\Delta m_1}{\Delta t}=\frac{\Delta m_2}{\Delta t}$$

$$\rho_1\,A_1\,\frac{\Delta x_1}{\Delta t}=\rho_2\,A_2\,\frac{\Delta x_2}{\Delta t}$$

Grenzübergang $\Delta t \rightarrow 0$:

$$\rho_1\,A_1\,\frac{\mathrm{d} x_1}{\mathrm{d} t}=\rho_2\,A_2\,\frac{\mathrm{d} x_2}{\mathrm{d}t}$$

und mit der Definition für Geschwingigkeitkeit erhalten wir die Kontinuitätsgleichung für den Massenstrom:

$$\rho_1\,A_1\,v_1=\rho_2\,A_2\,v_2$$

Wenn die Dichte sich nicht ändert, z.B. in inkompressiblen Flüssigkeiten/Gase, ist $\rho_1=\rho_2$ und es gilt

$$A_1\,v_1=A_2\,v_2=\mathrm{konst.}$$
