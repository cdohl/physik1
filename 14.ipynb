# 14. Reale Gase und Phasenumwandlung

* bisher: ideales Gas
  * kein Eigenvolumen der Teilchen
  * nur elastische Stöße zwischen und und mit der Gefäßwand
  ⇒ Druck $p$ an der Wand
  
  * Zustandsgleichung
  $$p\,V=n\,{\cal R}\,T$$
  $$p\,\frac{V}{n}={\cal R}\,T$$
  $$p\,V_m={\cal R}\,T$$
  
  $V_m$ molares Volumen, $\left[V_m\right]=\mathrm{1m^{3}\,mol^{-1}}$.
  * ideales Gas ist $V_m$ komprimierbar.
  
## 14.1 Zustandsgleichung des realen Gases (van der Waals-Gleichung)

$$\left(p+\frac{a}{V_m^2}\right)\,\left(V_m - b\right)={\cal R}\,T$$

$a,b$ sind gasspezifische Konstanten, $b$: molares Kovolumen, $\frac{a}{V_m^2}$: Binnendruck

* Gasteilchen hat ungefähr ein Eigenvolumen von 

  $$V_E=\frac{4\pi}{3}R^3$$
  
  $R$: Radius des Moleküls
  
  $$b=4\,N_A\,V_E$$
  
* Binnendruck  $\frac{a}{V_m^2}$

  <img src="figures/14.1_1.png" width=400px>
  
  
  * an der Gefäßwand wird der Druck $p$ gewesen 
  
  * im Inneren des Gases ist der Gesamtdruck größer als $p$ ⇒ $p+\frac{a}{V_m^2}$
  
* Van der Waals Gleichung

  $$p=p(V_m)=\frac{{\cal R}\,T}{V_m\,b}-\frac{a}{V_m^2}$$
  
  Unten sind ein paar Isothermen gezeichnet für $\mathrm{CO_2}$ berechnet. Die Konstanten $a$ und $b$ gibt es beispielsweise [hier](https://en.wikipedia.org/wiki/Van_der_Waals_constants_%28data_page%29).
  
  

```python
%matplotlib inline
%config InlineBackend.close_figures=False 

import ipywidgets as widgets
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['font.size'] = 15

Vm = np.linspace(0.00002, .002, 1000)

a=3.640*0.1
b=0.04267*1e-3
R=8.13
Tk=31.5+273.15
Ta=[.9* Tk, Tk, 2*Tk, 4*Tk]

plt.figure(1, figsize=(8, 6));
ax=plt.gca()
ax.clear()

for T in Ta:
    p=R*T/(Vm-b)-a/(Vm**2)
    ax.plot(Vm*1e3,p/1e5,label="$T=${0} K".format(T))
ax.set_ylabel('Druck $p$ (bar)')
ax.set_xlabel('Volume $V_m$ (L/mol)')
ax.set_ylim(-0.05, 200)
plt.legend()
plt.show();
```

 $$p(V_m)=\frac{{\cal R}\,T}{V_m\,b}-\frac{a}{V_m^2}$$

 $$\left.\frac{\mathrm{d}^2p}{\mathrm{d}{V_m}^2}\right|_{T_{Kr},p_{D,Kr},V_{m,Kr}}=0$$
 
 
 nach kurzer Rechnung:
 
 
 $$V_{m,Kr}=3\,b$$
 
 $$p_{D,Kr}=\frac{a}{27\,b^2}$$
 
 $$T_{Kr}=\frac{8a}{27 R\,b}$$
 
## 14.2 Clausius-Clapeyron-Gleichung und Dampfdruck

$$p_D(T)=p_D(T_0)\,\exp\left[\frac{r_m}{R}\,\left(\frac{1}{T_0}-\frac{1}{T}\right)\right]$$

* $r_m$: Molare Verdampfungswärme, die Energie, die notwendig ist die Flüssigkeit in den gasförmigen Zustand zu bringen. Ihr Ursprung liegt in der Anziehungskraft der Moleküle.  

* $T_0$: Bezugstemperatur

* $p_D(T_0)$: Dampfdruck bei der Bezugstemperatur

Eine Flüssigkeit siedet, wenn der auf der Flüssigkeitsoberfläche lastende Druck $p$ gleich dem Dampfdruck ist. 

Unten sehen Sie die Dampfdruckkurve von Wasser:

```python
%matplotlib inline
%config InlineBackend.close_figures=False 

import ipywidgets as widgets
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['font.size'] = 15

T = np.linspace(274, 450, 1000)
#$$p_D(T)=p_D(T_0)\,\exp\left[\frac{r_m}{R}\,\left(\frac{1}{T_0}-\frac{1}{T}\right)\right]$$
rm=40.8e3
R=8.13
pd0=1.013e5
T0=273.15+100.



plt.figure(1, figsize=(8, 6));
ax=plt.gca()
ax.clear()

p=pd0*np.exp(rm/R*(1./T0-1./T))
ax.plot(T-273.15,p/1e5)
ax.set_ylabel('Druck $p$ (bar)')
ax.set_xlabel('Temperatur $T$ ($^\circ$C)')
#ax.set_ylim(-0.05, 200)
plt.show();
```
