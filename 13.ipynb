# 13. Kreisprozesse
## 13.1 Wärme-Kraft- und Kraft-Wärme-Maschine

*Kreisprozess*: Abfolge von Zustandsänderungen zum Ausgangspunkt zurück.

<img src="figures/13.1_1.png" width=400px>

Da die Endtemperatur = Ausgangstemperatur ist die Änderung der innneren Energie für einen vollen Umlauf 0!

$$\oint\mathrm{d}U=0$$

* Vier Zustandsgrößen:
  $p, V, T, U$
  
* zwei Prozessgrößen: $Q, W$, diese sind abhängig vom gewählten Weg der Zustandsänderung.

* Ziele von Maschinen:
  * Umwandlung von Wärme in mechanische Energie
    ⇒ Wärme-Kraft-Maschine
    
  * Transport von Wärme durch mechanische Arbeit
    ⇒ Kraft-Wärme-Maschine

* Wärme-Kraft-Maschine
  
  $$T_h>T_t$$


  <img src="figures/13.1_2.png" width=250px>
  
  da $\Delta U=0$ ist
  $$Q_{zu}=\left|Q_{ab}+W_{ges}\right|$$
  
  Rechtsläufig (im Uhrzeigersinn im $pV$-Diagramm).
  
  Wirkungsgrads:
  
  $$\eta=\frac{\left|W_{ges}\right|}{Q_{zu}}$$
  
* Kraft-Wärme-Maschine
  
  linksläufig (gegen den Uhrzeigersinn im $pV$-Diagramm)
  
  <img src="figures/13.1_3.png" width=250px>
  
  $$Q_{zu}+W_{ges}=|\underbrace{Q_{ab}}_{negativ}|$$
  
  $\epsilon$: Leistungskennzahl
  
  Kältemaschine:
 
  
  $$\epsilon_{KM}=\frac{Q_{zu}}{W_{ges}}$$
  
  $Q_{zu}$ dem kalten Reservoir entzogene Wärme.
  
  Wärmepumpe:
  
  
  $$\epsilon_{WP}=\frac{|Q_{ab}|}{W_{ges}}$$
  
  $Q_{ab}$ dem Reservoir zugeführte Wärme.


## 13.2 Der Carnot-Prozess

#### Beispiel Wärme-Kraft-Maschine

 <img src="figures/13.1_4.png" width=450px>
 
 * A → B isotherme Expansion bei hoher Temperatur $T_h$
 * B → C adiabatische Expansion 
 * C → D isotherme Kompression bei Temperatur $T_t$
 * D → A adiabatische Kompression
 
 * Wirkungsgrad
 
 $$\eta=\frac{|W_{ges}|}{Q_{zu}}$$
 
 $$W_{ges}=W_{AB}+W_{BC}+W_{CD}+W_{DA}$$
 
 $$Q_{zu}=Q_{AB}$$
 
 $$Q_{ab}=Q_{CD}$$
 
 
 
### A → B

<img src="figures/13.1_5.png" width=400px>

Isotherm: $T=\mathrm{konst.}=T_h$

$$p=\frac{n\,{\cal R} T_h}{V}$$

Bestimmung von $W_{AB}$:

\begin{eqnarray}
W_{AB}&=&-\int_{V_A}^{V_B} p \mathrm{d}V\\
&=&-\int_{V_A}^{V_B} \frac{n\,{\cal R} T_h}{V} \mathrm{d} V\\
W_{AB}&=&-n\,{\cal R}\,T_h\,\ln\left(\frac{V_B}{V_A}\right)<0
\end{eqnarray}


Bestimmung von $Q_{AB}$:

Da $T=T_h=\mathrm{konst.}$ ist $\mathrm{d}T=0$ und daher $\mathrm{d}U=0$.

Wir benutzen den 1. H.S.:

\begin{eqnarray}
\underbrace{0}_{\mathrm{d}U}&=&\mathrm{d}Q+\mathrm{d}W\\
\mathrm{d}Q&=&-\mathrm{d}W\\
Q_{AB}&=&-W_{AB}\\
Q_{AB}&=&Q_{zu}=n\,{\cal R}\,T_h\,\ln\left(\frac{V_B}{V_A}\right)
\end{eqnarray}


### C → D

Isotherm: $T=\mathrm{konst.}=T_t$

$$W_{CD}=-n\,{\cal R}\,T_h\,\ln\left(\frac{V_D}{V_C}\right)>0$$

$$Q_{CD}=Q_{ab}=-W_{CD}=n\,{\cal R}\,T_h\,\ln\left(\frac{V_D}{V_C}\right)$$


### B → C

Adiabatisch: $\mathrm{d}Q=0$

1. H.S. $\mathrm{d}U=\mathrm{d}W$

$$n\,C_{M,V}\,\mathrm{d}T=\mathrm{d}W$$

$$n\,C_{M,V}\,(T_t-T_h)=W_{BC}<0$$

### D → A

Adiabatisch: $\mathrm{d}Q=0$

$$n\,C_{M,V}\,(T_h-T_t)=W_{DA}>0$$

$$W_{BC}+W_{DA}=0$$


* $W_{ges}$

  $$W_{ges}=W_{AB}+W_{CD}<0$$
  
  $$|W_{ges}|=n\,{\cal R}\,T_h\,\ln\left(\frac{V_B}{V_A}\right)+n\,{\cal R}\,T_t\,\ln\left(\frac{V_D}{V_C}\right)$$
  
  $$Q_{zu}=Q_{AB}=n\,{\cal R}\,T_h\,\ln\left(\frac{V_B}{V_A}\right)$$
  
  $$\eta_C=\frac{|W_{ges}|}{Q_{zu}}=1+\frac{T_t\,\ln\left(\frac{V_D}{V_C}\right)}{T_h\,\ln\left(\frac{V_B}{V_A}\right)}\tag{13.2.1}$$
  
  Nach kurzer Rechnung kann gezeigt werden, dass $V_A\,V_C=V_B\,V_D$. Damit lässt sich Gl. (13.2.1) vereinfachen, da
  
  $$\frac{V_D}{V_C}=\left(\frac{V_B}{V_A}\right)^{-1}$$
  
  und wir erhalten:
  
  $$\eta_C=1-\frac{T_t}{T_h}\quad .$$
  
* hoher Wirkungsgrad wenn

  * niedriges $T_t$
  
  * hohes $T_h$
  
#### Beispiel Kraft- Wärme-Maschine

<img src="figures/13.1_7.png" width=500px>
* Der Zyklus ist nun A→D→C→B→A...
* $Q_{BA}=-Q_{AB}$
* $W_{ges}$ wird hier zugeführt


* Kältemaschine
  Leistungszahl
  $$\epsilon_{KM}=\frac{Q_{zu}}{W_{ges}}=\frac{T_t}{T_h-T_t}$$
  
  Abschätzung:
  
  $T_h=300\,$K und $T_t=270\,$K
  
  $$\epsilon_{KM}=\frac{270\,\mathrm{K}}{30\,\mathrm{K}}=9$$
  
* Wärmepumpe
  Leistungszahl
  $$\epsilon_{WP}=\frac{|Q_{ab}|}{W_{ges}}=\frac{T_h}{T_h-T_t}$$
  $$\epsilon_{WP}=\eta^{-1}$$
  $$\epsilon_{WP}=\epsilon_{KM}+1$$
  Abschätzung:
  
  $T_h=300\,$K und $T_t=270\,$K
  
  $$\epsilon_{KM}=\frac{300\,\mathrm{K}}{30\,\mathrm{K}}=10$$
  
 ## 13.3 Der Stirling-Prozess (Heißluftmotor)
 
<img src="figures/13.1_8.png" width=500px>

* A→B isotherme bei $T_h$
* B→C isochor
* C→D isotherme bei $T_t$
* D→A isochor 

* Bei allen vier Teilprozessen erfolgt Wärmeaustausch

  $$Q_{BC}=-Q_{DA}$$

  $Q_{ab}^*$ wird hier zwischengespeichert und für $Q_{zu}^*$ wieder verwendet.
  
  * Mit dieser Wärmeregeneration ist der Wirkungsgrad 
  
    $$\eta_{stir}=\frac{|W_{ges}|}{Q_{AB}}=1-\frac{T_t}{T_h}$$
  
  * ohne Rückführung der Wärme
  
    $$\eta_{ohne}=\frac{|W_{ges}|}{Q_{AB}+Q_{DA}}<\eta_{stir}$$

## 13.4 Reversible und Irreversible Prozesse
* Eine große Anzahl von Prozessen sind zwar mit dem 1. H.S. (Energiesatz) vereinbar, werden aber experimentell nie beobachtet.

* Prozesse in abgeschlossenen Systemen laufen immer in Richtung des thermischen Gleichgewichtes

* ohne äußere Einwirkung laufen diese Prozesse nie in umgekehrter Richtung, sie heißen irreversibel

* in der Natur verlaufen alle Prozesse irreversibel (Expansion eines Gases ins Vakuum, Vorgänge, bei denen Wärme freigesetzt wird wie Oxidation, Lösung eines Stoffes und Diffusion, inelastische Verformung)

* Systeme befinden sich bei irreversiblen Zustandsänderungen nicht im Gleichgewischt

* Reversibel Prozesse: System verlässt das Gleichgewicht nicht, sie sind ohne äußere Einflüsse umkehrbar

## 13.5 2. Hauptsatz der Thermodynamik

Es gibt keine periodisch arbeitende Maschine, die einen höheren Wirkungsgrad zeigt als die Carnot-Maschine (bei gleichen Temperaturen).

$$\eta_{Real}\le 1-\frac{T_t}{T_h}=\eta_{Carnot}$$

Es gibt keine periodisch arbeitende Maschine, die Arbeit erzeugt und dabei nur einem Wärmereservoir Wärme entzieht.


## 13.6 Entropie

* Carnot-Prozess (ideal) ist reversibel

  $$\eta_C=\frac{Q_{AB}+Q_{CD}}{Q_{AB}}=1-\frac{T_t}{T_h}$$
  
  umgeformt ergibt sich
  
  $$\frac{Q_{AB}}{T_h}+\frac{Q_{CD}}{T_t}=0$$
  
  $Q_{AB}$ ist die bei hoher Temperatur ausgetauschte Wärme.
  
  $Q_{CD}$ ist die bei tiefer Temperatur ausgetauschte Wärme.
  
  
* Verallgemeinerung

  $$\sum_i^N\frac{Q_{i,rev}}{T_i}=0$$

* Clausius: 

   * neue Zustandsgröße Entropie $S$
   
   * die Änderung von $S$ wird durch die reversibel ausgetauschte Wärmemengen bei den entsprechenden Temperaturen bestimmt.
   
     $$\Delta S=\frac{Q_{rev}}{T}$$

* für infinitesimal kleine Änderungen

  $$\mathrm{d}S=\frac{\mathrm{d}Q_{rev}}{T}$$
  
  $$\mathrm{\left[S\right]=\frac{J}{K}}$$
  
* für alle reversiblen Prozess gilt, dass das integral über einen geschlossenen Weg

  $$\oint\frac{\mathrm{d}Q_{rev}}{T}=0$$
  
  ist.

* in der Natur: sich selbst überlassene Systeme streben dem Zustand größter Entropie zu $\Delta S >0$.

* statistische Bedeutung von $S$ 
  $$S=k\,\ln W$$
  $k$ ist die Boltzmann Konstante und $W$ die thermodynamische Wahrscheinlichkeit.
