# 6. Arbeit und mechanische Energie

## 6.1 Arbeit

Eine Kraft $\vec{F}$, die einen Körper eine bestimmte Wegstrecke $\mathrm{d}\vec{r}$ verschiebt, verrichtet eine Arbeit:

$$\mathrm{d}W=\vec{F}\cdot\mathrm{d}\vec{r}$$

Die Gesamtarbeit bei einer Verschiebung

$$\int_0^{W_{12}}=\int_{\vec{r}_1}^{\vec{r}_2}\vec{F}\cdot\mathrm{d}\vec{r}$$

$$W_{12}=\int_{\vec{r}_1}^{\vec{r}_2}\vec{F}\cdot\mathrm{d}\vec{r}\tag{6.1.1}$$

Allgemeine Abkürzung für Arbeit $W$, die Einheit ist $\left[W\right]=\mathrm{J}$ und steht für Joule.

Bedenken Sie, dass Sie ein Skalarprodukt hier haben:
$$ \vec{F}\cdot\mathrm{d}\vec{r}=|\vec{F}|\,|\mathrm{d}\vec{r}|\,\cos (\vec{F},\mathrm{d}\vec{r})$$

Die Arbeit ist maximal: $\vec{F}\parallel\mathrm{d}\vec{r}$

Die Arbeit ist minimal: $\vec{F}\perp\mathrm{d}\vec{r}$

Ein Beispiel für minimale Arbeit $W=0$ ist die Radialkraft $\perp$ $\vec{v}$.

<b>Verschiebungsarbeit (gegen die Reibungskraft)</b>

<img src="figures/6.1_1.png" width=300px>

$$\vec{v}=\mathrm{konst.}=v_x\,\vec{e}_x$$

$$\mathrm{d}\vec{r}\parallel\vec{e}_x$$

$$\vec{F}_{Zug}=F_x\,\vec{e}_x+F_y\,\vec{e}_y$$

\begin{eqnarray}
F_x&=&\left|\vec{F}_{Zug}\right|\,\cos\beta\\
F_y&=&\left|\vec{F}_{Zug}\right|\,\sin\beta
\end{eqnarray}

Wenn $\vec{v}=\mathrm{konst.}$ dann gilt

$$\vec{F_x}\,\vec{e}_x+\vec{F}_{GL}$$
$$\vec{F_x}\,\vec{e}_x=-\vec{F}_{GL}$$

und der Weg des Körpers ist

$$\mathrm{d}\vec{r}=\vec{e}_x\,\mathrm{d}x$$

Dann können wir das Integral Gl. (6.1.1) schreiben


\begin{eqnarray}
W_{12}&=&\int_{x_1}^{x_2}\vec{F}_{Zug}\cdot\mathrm{d}\vec{r}\\
&=&\int_{x_1}^{x_2}\left(F_x\,\vec{e}_x+F_y\,\vec{e}_y \right)\vec{e}_x\,\mathrm{d}x\\
&=&\int_{x_1}^{x_2} F_x\,\vec{e}_x\cdot\vec{e}_x\,\mathrm{d}x + \int_{x_1}^{x_2} F_y\,\underbrace{\vec{e}_y\cdot\vec{e}_x}_{\vec{e}_x\cdot\vec{e}_y=0}\,\mathrm{d}x
\end{eqnarray}

Da $\vec{e}_x \perp \vec{e}_y$ ist

$$W_{12}=\int_{x_1}^{x_2}F_{x}\,\mathrm{d}x$$

und da $F_x=\mathrm{konst.}$ erhalten wir

$$W_{12}=F_x\,\left(x_2-x_1\right)\quad .$$

Nur die Komponente in $\vec{e}_x$ von $\vec{F}_{Zug}$ verrichtet Arbeit.

<b>Hubarbeit (gegen die Gravitation im erdnahen Raum)</b>


*Erdnah* bedeutet, dass die Gewichtskraft konstant ist. Dies ist sie im allgemeinen nicht. Wir werden später ein verallgemeinertes Gesetz für die Massenanziehung kennenlernen.


<img src="figures/6.1_2.png" width=300px>

Die Kraft die notwendig ist um die Masse zu Heben ist 

\begin{eqnarray}
\vec{F}_{Hub}&=&-\vec{F}_g\\
&=&m \, g\, \vec{e}_z
\end{eqnarray}

Dies geschieht ohne Beschleunigung, also "unendlich" langsam.

$$W_{12}=\int_0^h \vec{F}_{Hub}\cdot\mathrm{d}\vec{r}=
\int_0^h m\,g\,\underbrace{\vec{e}_z\cdot\vec{e}_z}_{=1}\,\mathrm{d}z$$

und wir erhalten

$$W_{12}=m\,g\,h\quad .$$

Das gilt nur für konstante Masse $m$ und konstantes $g$.


<b>Beschleunigungsarbeit</b>

* beschleunigende Kraft: $\vec{F}_B=m\,\vec{a}$
  $$\vec{F}_B=m\,\frac{\mathrm{d}\vec{v}}{\mathrm{d}t}$$
  
$$W_B=W_{12}=\int_{\vec{r}_1}^{\vec{r}_2} m\,\frac{\mathrm{d}\vec{v}}{\mathrm{d}t} \cdot \mathrm{d} \vec{r}=\int_{\vec{r}_1}^{\vec{r}_2} m\,\frac{\mathrm{d}\vec{r}}{\mathrm{d}t} \cdot \mathrm{d} \vec{v}$$

weil 

$$\frac{\mathrm{d}\vec{r}}{\mathrm{d}t}=\vec{v}$$

schreiben wir das Integral

\begin{eqnarray}
W_{12}&=&\int_{\vec{v}_1(\vec{r}_1)}^{\vec{v}_2(\vec{r}_2)}m\,\vec{v}\cdot\mathrm{d}\vec{v}\\
&=&\left.\frac{m}{2}\vec{v}^2\right|_{\vec{v}_1}^{\vec{v}_2}=
\frac{m}{2}\left({\vec{v}_2}^2-{\vec{v}_1}^2\right)
\end{eqnarray}

Damit ist die Beschleunigungsarbeit für einen Körper von Geschwindigkeit $\vec{v}_1$ nach Geschwindigkeit $\vec{v}_2$ 

$$W_B=\frac{m}{2}\left({\vec{v}_2}^2-{\vec{v}_1}^2\right)$$



## 6.2 Leistung

### Momentane Leistung

$$P=\frac{\mathrm{d}W}{\mathrm{d}t}$$

Einheiten: $\left[P\right]=1\,$W 


\begin{eqnarray}
d\mathrm{W}&=&P\,\mathrm{d} t\\
W_{12}&=&\int_{t_1}^{t_2}\,P\,\mathrm{d}t
\end{eqnarray}

<img src="figures/6.2_1.png" width=400px>

Die mittlere Leistung $\bar{P}$ im Zeitinterval $\Delta t$ erhält man duch

$$\bar{P}=\frac{W_{12}}{\Delta t}=\left< P\right>$$


## 6.3 Mechanische Energien

### 6.3.1 Kinetische Energie

* Zustand von Körpern $\rightarrow$ Energie wird zugeordnet
* hängt von der Geschwindigkeit ab
* $E_{kin}=\frac{m}{2} v^2$
* Kinetische Energie kann durch Beschleunigungsarbeit geändert werden

  \begin{eqnarray}
  E_{kin}^{(2)}&=&\frac{m}{2} v_2^2\\
  E_{kin}^{(1)}&=&\frac{m}{2} v_1^2\\
  \Delta E_{kin}=E_{kin}^{(2)}-E_{kin}^{(1)}&=&\frac{m}{2}\left(v_2^2-v_1^2\right)
  \end{eqnarray}
  
  Die Differenz in der kinetischen Energie $\Delta E_{kin}=W_{12}$ ist die Beschleuningungsarbeit.
  
### 6.3.2 Potentielle Energie

* Ist gespeicherte, die wieder freigesetzt werden kann
* Definition der potentiellen Energie
  $$E_{pot}=-\int_{\vec{r}_0}^{\vec{r}} \vec{F_{kon}}\,\mathrm{d}\vec{r}$$
  $\vec{r}_0$ ist der Nullpunkt und kann z.B. mit der potentiellen Eenergie von Null besetzt werden, d.h. $E_{pot}(\vec{r}=\vec{r}_0)=0$. Die Kraft im Integral muss konservativ sein um ihr eine potentielle Energie zuordnen zu können, z.B. Federenergie:
  
  
  
* $E_{pot}$ für Feder: $E_{pot}(x=0)=0$

  \begin{eqnarray}
  E_{pot}&=&-\int_0^{x_2}\vec{F}_F\cdot\mathrm{d}\vec{r}\\
  &=&-\int_0^{x_2}-k\,x\,\vec{e}_x\cdot\vec{e}_x\,dx\\
  E_{pot}&=&\frac{k}{2}\,x_2^2
  \end{eqnarray}
  
* $E_{pot}$ für Heben eines Körpers auf Höhe $h$
  <img src="figures\6.2_2.png" width=300px>

  \begin{eqnarray}
  E_{pot}&=&-\int_0^{h}\vec{F}_G\cdot\mathrm{d}\vec{r}\\
  &=&-\int_0^{h}-m\,g\,\vec{e}_z\cdot\vec{e}_z\,dz\\
  E_{pot}&=&m\,g\,h
  \end{eqnarray}

### 6.3.3 Energieerhaltungssatz der Mechanik

* Mechanische Energie ($E_{kin}$ und $E_{pot}$) können ineinander umgewandelt werden.

⇒ die Zunahme von $E_{kin}$ aus dem Vorrat von $E_{pot}$

⇒ die zeitliche Änderung sind entgegengesetzt

$$\frac{\mathrm{d}E_{kin}}{\mathrm{d}t}=-\frac{\mathrm{d}E_{pot}}{\mathrm{d}t}$$

$$\frac{\mathrm{d}\left(E_{kin}+E_{pot}\right)}{\mathrm{d}t}=0$$

Oder anders ausgedrückt: 

Für abgeschlossene Systeme gilt die Energieerhaltung:

$$E_{kin}+E_{pot}=\mathrm{konst.}=E_{ges}$$
