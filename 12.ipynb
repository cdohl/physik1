## 12.1 Absolute Temperatur

* für ein Gas mit $N$ Teilchen ⇒ Innere Energie

  $$ U=n\frac{f_{ges}}{2}\,{\cal R}\,T$$
 
  für den Anteil der Translation
  
  $$U_T=N\,\overline{E_{kin}}=N\,\frac{3}{2}\, k\, T$$
  $$ = N \frac{1}{2}\mu\,\overline{v^2}$$
  
  $$\overline{v^2}\propto T$$
  
  Die absolute Temperatur ist ein Maß für die mittlere kinetische Energie der Teilchen.
  
  Wenn $\overline{v^2}\rightarrow 0$ dann $T\rightarrow 0$.
  
* Einheit der Temperatur $[T]=1\,$K

* Es gibt andere relative Temperaturskalen
  * Celsius Skala
  * $\rightarrow$ Gefrierpunkt des Wasser $\vartheta_G=0\,^\circ$C
  * $\rightarrow$ Siedpunkt des Wasser $\vartheta_S=100\,^\circ$C
  
* Umrechnung 
  $$\vartheta/^\circ\mathrm{C}=T/\mathrm{K}-273.15$$
  $$0^\circ\mathrm{C}=273.15\,\mathrm{K}$$
  $$-273.15^\circ\mathrm{C}=0\mathrm{K}$$

* isobare Zustandsänderung $p=\mathrm{konst.}$
  ⇒ Messung des Volumens in Abhängigkeit von der Temperatur
  
  <img src="figures/12.1_1.png" width=270px>
  
  Wann nimmt ein ideales Gas das Volumen von $V=0$ ein: $\vartheta=-273.15\,^\circ$C, $T=0\,$K.
  
  
Beispiel eines sehr kalten Gases (Erklärung in der Vorlesung):

## 12.2 Wärme und Wärmekapazität

* Wärme $Q$ ist eine Übertragungsform von Energie

* Welche Wärmemenge ⟷ Energiemenge muss man einem Körper zuführen, um dessen Temperatur um $1\,$K zu ändern?

* Wärmekapazität:

  $$C=\frac{\mathrm{d}Q}{\mathrm{d}T}$$

  oder 
  
  $$\mathrm{d}Q=C\,\mathrm{d}T$$
  
  für endliche Temperaturdifferenz $\Delta T$
  
  $$C=\frac{\Delta Q}{\Delta T}$$
  
  Einheit der Wärmekapazität: $[C]=1\,$J$\,$K$^{-1}$

* *nur für* Vorgänge ohne Volumenänderung

  $$\mathrm{d}Q=\mathrm{d}U\,\propto\, \mathrm{d}T$$
  
* spezifische Wärmekapazität
  
  $$c=\frac{1}{m}\,\frac{\Delta Q}{\Delta T}$$
  
  $m$: Masse des Stoffes, $m=\mu\,N$
  
  Einheit der spezifischen Wärmekapazität: $[c]=1\,$J$\,$K$^{-1}\,$kg$^{-1}$

* molare Wärmekapazität

  $$C_M=\frac{1}{n}\frac{\Delta Q}{\Delta T}$$
  
  $$C_M=c\,M$$
  
  $M$: Molare Masse
  
  Einheit der molaren Wärmekapazität: $[C_M]=1\,$J$\,$K$^{-1}\,$mol$^{-1}$
  
## 12.3 0. Hauptsatz der Thermodynamik

Körper, die miteinander im Kontakt stehen (wechselwirken), streben dem thermodynamischen Gleichgewicht zu.

⇒  Temperaturausgleich

$$T_{K1}=T_{K2}=T_{K3}$$

Wäremeaustausch als Wechselwirkung.

## 12.4 Formulierung des Energiesatzes (1. Hauptsatz)

* ideale Gase
* Zustandsgleichung: $p\,V=n\,{\cal R}\,T$
* Innere Energie: $U=n\frac{1}{2}\,f_{ges}\,{\cal R}\,T$

Wie kann man bei einem Gas die innere Energie, d.h. die Temperatur ändern?

  1. Zuführen von Wärme
  2. Komprimierung des Gases, mechanische Arbeit wir dem Gas zugeführt

* Mathematische Formulierung des 1. Hauptsatzes

  $$\mathrm{d}U=\mathrm{d}W+\mathrm{d}Q$$
  
  Änderung der inneren Energie durch die Zu- und Abführung von Arbeit ($\mathrm{d}W$) oder Wärme ($\mathrm{d}Q$).

* Alles, was dem Gas von außen zugeführt wird, soll positiv sein:

  * $\mathrm{d}Q>0$ Wärmezufuhr von außen
  * $\mathrm{d}Q<0$ Wärmeabgabe nach außen
  * $\mathrm{d}W>0$ Komprimierung des Gases ⇒ Volumen nimmt ab, weil $\mathrm{d}V<0$ ⇒ $\mathrm{d}W=-p\,\mathrm{d}V$
  * $\mathrm{d}W>0$ und umgekehrt, Expansion des Gases $\mathrm{d}V>0$
  
* 1. Hauptsatz der Thermodynamik

  $$\mathrm{d}U=-p\,\mathrm{d}V+\mathrm{d}Q$$
  

⇒ Zustandsgleichung + 1. HS führt zu 4 möglichen Zustandsänderungen

* isochor, $V=\mathrm{konst.}$, ($\mathrm{d}V=0$)
* isobar, $p=\mathrm{konst.}$, ($\mathrm{d}p=0$)
* isotherm, $T=\mathrm{konst.}$, ($\mathrm{d}T=0$, $\mathrm{d}U=0$)
* adiabatisch, $Q=\mathrm{konst.}$, ($\mathrm{d}Q=0$) aber $p, V, T$ ändern sich. Kein Wärmeaustausch mit der Umgebung.

## 12.5 Wärmekapazität des idealen Gases

* molare Wärmekapazität

  $$C_M=\frac{1}{n}\frac{\mathrm{d}Q}{\mathrm{d}T}$$

  a) isochore Zustandsänderung
     $$\mathrm{d}V=0$$
     ⇒ 
     $$\mathrm{d}W=0$$
     und damit keine Arbeitsverrichtung.
     
     Wegen des 1. HS ⇒ $\mathrm{d}U=\mathrm{d}Q$
     $$C_{M,V}=\frac{1}{n}\left.\frac{\mathrm{d}Q}{\mathrm{d}T}\right|_{V=\mathrm{konst.}}$$
     
     $$C_{M,V}=\frac{1}{n}\frac{\mathrm{d}U}{\mathrm{d}T}\quad \leftarrow\quad
     \frac{\mathrm{d}U}{\mathrm{d}T}=n\,\frac{f_{ges}}{2}{\cal R}$$
     
     
  b) isobare Prozesse
  
  $$C_{M,P}=\frac{1}{n}\left.\frac{\mathrm{d}Q}{\mathrm{d}T}\right|_{p=\mathrm{konst.}}$$
  
  Mit dem 1. HS $$\mathrm{d}Q=p\,\mathrm{d}V+\mathrm{d}U$$
  
  $$C_{M,P}=\underbrace{\frac{1}{n}\frac{\mathrm{d}U}{\mathrm{d}T}}_{C_{M,V}}+
            \frac{1}{n}\,p\,\frac{\mathrm{d}V}{\mathrm{d}T}\tag{12.4.1}$$
            
  Für isobare Zustandsänderung ergibt sich aus dem idealen Gasgesetz:
  
  $$\frac{\mathrm{d}V}{\mathrm{d}T}=n\,\frac{{\cal R}}{p}$$
  
  Damit wird aus Gl. (12.4.1) 
  
  $$C_{M,P}=C_{M,V}+{\cal R}$$
  ⇒ es gilt immer $C_{M,P}>C_{M,V}$
  
  mit ${\cal R}=C_{M,P}-C_{M,V}$
  
  
  $$C_{M,P}=\frac{f_{ges}}{2}\,{\cal R}+ {\cal R}=\frac{f_{ges}+2}{2}\,{\cal R}$$

## 12.6 Anwendung des 1. Hauptsatzes auf Zustandsänderungen des idealen Gases

### 12.6.1 Isochore Zustandsänderung

Da $V=V_0=\mathrm{konst.}$ erhalten wird für das ideale Gasgesetz $p\,V_0=n\,{\cal R} T$ folglich $p=\frac{n\,{\cal R}T}{V_0}$.

Weiterhin ist aus $\mathrm{d}V=0$ ⇒ $\mathrm{d}W=\mathrm{d}Q$

Die Änderung der inneren Energie $\Delta U$ folgt allein aus der Temperaturänderung.

$$\Delta U=U_2-U_1=n\,C_{M,V}\left(T_2-T_1\right)$$

<img src="figures/12.5_1.png" width=600px>

### 12.6.2 Isobare Zustandsänderung

Da $p=p_0=\mathrm{konst.}$ erhalten wir für das ideale Gasgesetz $p_0\,V=n\,{\cal R}T$

$$V=\frac{n{\cal R} T}{p_0}$$

<img src="figures/12.5_2.png" width=600px>

* Berechnung der Arbeit $W_{12}$

$$\mathrm{d}W=-p\mathrm{d}V$$

$$\int_0^{W_{12}}\mathrm{d}W=\int_{V_1}^{V_2}-p\mathrm{d}V$$ mit $p=p_0=\mathrm{konst.}$

$$W_{12}=-p_0\int_{V_1}^{V_2}\mathrm{d}V$$

$$W_{12}=-p_0\left(V_2-V_1\right)V$$

Für unser Beispiel ist $V_2>V_1$ (Expansion) und $(V_2-V_1)>0$, damit ist $W_{12}<0$.
Mit unsere Definition gibt das Gas Arbeit nach außen ab.

* Veränderung der inneren Energie $\mathrm{d}U$

$$T_1=\frac{V_1 p_0}{n {\cal R}}\quad T_2=\frac{V_2 p_0}{n {\cal R}}$$

\begin{eqnarray}
Q_{12}&=&n\,C_{M,p}\left(T_2 - T_1\right)\\
      &=&n\,C_{M,p}\frac{p_0}{n{\cal R}}\left(T_2-T_1\right)
\end{eqnarray}

### 12.6.3 Isotherme Zustandsänderung

$$T=T_0=\mathrm{konst.}\quad p=\frac{n{\cal R}T_0}{V}$$

<img src="figures/12.5_3.png" width=600px>


* Bestimmung der Arbeit $W_{12}$ 
  
  Ideales Gas
  $$p=\frac{n {\cal R} T_0}{V}$$

  \begin{eqnarray}
  W_{12}&=&-\int_{V_1}^{V_2} p\,\mathrm{d}V\\
  &=&-n\,{\cal R}\,T_0\int_{V_1}^{V_2} \frac{\mathrm{d}V}{V}\\
  &=&-n\,{\cal R}\,T_0\,\left.\ln V\right|_{V_1}^{V_2}\\
  W_{12}&=&-n\,{\cal R}\,T_0\,\ln\left(\frac{V_2}{V_1}\right)
  \end{eqnarray} 

  Wenn $\left(\frac{V_2}{V_1}\right)>1$ ⇒  $\ln\left(\frac{V_2}{V_1}\right)>0$ und $W_{12}<0$.

  ⇒ das Gas expandiert und gibt Arbeit ab.

* $\mathrm{d}U=0$ da $\mathrm{d}T=0$
  nach dem 1. HS ist $\mathrm{d}Q=-\mathrm{d}W$
  
  Die dem Gas zugeführte Wärmemengen $Q_{12}=-W_{12}$, $\mathrm{d}Q>0$. Dem Gas zugeführte Wärmemenge wird vollständig als Arbei wieder abgegeben.
  
<img src="figures/12.5_4.png" width=350px>  


### 12.6.4 Adiabatische Zustandsänderung

* $\mathrm{d}Q=0$ kein Wärmeaustausch mit der Umgebung
  * sehr schnelle Vorgänge
  * Wärmeisolation
  
* 1. HS  $\mathrm{d}U=\mathrm{d}W$

  $$n \,C_{M,V}\mathrm{d}T=-p\,\mathrm{d}V$$
  
  $$n \,C_{M,V}\mathrm{d}T= -n\,{\cal R}\, T \frac{\mathrm{d}V}{V}$$
  
  Trennung der Variablen
  
  $$\frac{\mathrm{d}T}{T}=-\frac{{\cal R}}{C_{M,V}}\,\frac{\mathrm{d}V}{V}\tag{12.6.1}$$
  und der Idendität ${\cal R}=C_{M,P}-C_{M,V}$, bzw. 
  $$\frac{\cal R}{C_{M,V}}=\frac{C_{M,P}}{C_{M,V}}-1=\kappa-1$$
  
  Hier wurde der Adiabatenexponent $\kappa$ eingeführt. Er ist eine Funktion der Freiheitsgrade $f_{ges}$
  
  $$\kappa=\frac{f_{ges}+2}{f_{ges}}$$
  
  Gleichung (12.6.1) kann leicht integriert werden und wir erhalten
  
  $$\frac{T_2}{T_1}=\left(\frac{V_1}{V_2}\right)^{\kappa-1}$$
  
  In adiabatischen Prozessen sind die Variablen $T$, $p$ und $V$ durch die sogenannten Poisson Gleichungen verbunden:
  
  \begin{eqnarray}
  T\,V^{\kappa -1}=\mathrm{konst.}\\
  p\,V^\kappa=\mathrm{konst.}\\
  T^{\kappa}p^{1-\kappa}=\mathrm{konst.}
  \end{eqnarray}
